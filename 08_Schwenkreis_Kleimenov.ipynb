# Introduction

Gamma-ray bursts (GRB's) are some of the most powerful astrophysical phenomena observed in the universe and are characterized by brief but very intense flashes of gamma-ray radiation, typically lasting for less than a minute. These events were first detected in the 1960's by the Vela satellites, and can be seen up to a few times a day at random locations in the sky. They take place at cosmological distances, so located well outside of the local group of galaxies.

GRB's are classified in two subcategories, **short-GRB's** with a duration below $2s$ and **long-GRB's** with a duration above $2s$. This is motivated by the histogram of GRB Observations, which peaks at $\sim 0.3s$ and $\sim 30s$, and the peaks are seperated by a dip at $2s$. The origin of short- and long-GRB's is still not fully understood, but it has been shown that short-GRB's can occur from the merger of compact objects in binary systems and that long-GRB's are a possible outcome of the collapse of massive starts (mass $\gtrapprox 15 M_\odot$).

The energy released in such bursts is between $10^{48}$ and $10^{55}$ erg if emitted isotropically, meaning the strongest bursts are able to rival the luminosity of the entire universe. In reality, this energy is released in highly relativistic jets and highly anisotropically, with jet opening angles in the range of $\sim 2 - 10$ degrees. This limits the released energy to around $10^{52}$ erg.

Additionally, these bursts produce **afterglows**, which are long lived emissions in a broad spectrum, arising from the interaction of the ejected material with the surrounding interstellar medium. These afterglows and the associated lightcurves are the main topic of interest in this project and will be investigated using the physics of shockwaves combined with a simple one-zone model approach to determine the kinematics and from this the spectral emission. # **1. Relativistic hydrodynamics**

## 1.1. The energy-momentum tensor of a relativistic fluid

To construct the equations of relativistic hydrodynamics, one requires to know the energy-momentum tensor of a relativistic fluid.

The standard energy-momentum tensor, known from classical electrodynamics course, can be read as following:
$$ T = \begin{pmatrix}
    e & p_x & p_y & p_z \\
    p_x & P & \sigma_{xz} & -\sigma_{xy} \\
    p_y & -\sigma_{xz} & P & \sigma_{xz} \\
    p_z & \sigma_{xy} & -\sigma_{xz} & P
\end{pmatrix},\quad(1.1)
$$
where $e$ is the total energy density, $\mathbf{p}$ is momentum-density, P is pressure and $ \hat\sigma $ corresponds to viscous forces tensor.

In a reference frame where the considered fluid is at rest, its momentum density $\mathbf{p} = 0$, and also the Pascal's law is valid, which reads $ T^{ij}d\mathbf{f}_i = Pd\mathbf{f}_j, $ and gives $T^{ij} = \delta^{ij}. $

Hence, in this very frame, the tensor $(1.1)$ simplifies to $(1.2)$.
$$ T = \begin{pmatrix}
    e & 0 & 0 & 0 \\
    0 & P & 0 & 0 \\
    0 & 0 & P & 0 \\
    0 & 0 & 0 & P
\end{pmatrix}\quad(1.2)
$$

Introducing the 4-velocity $u^\mu = (\Gamma, \Gamma\mathbf{v}/c)$, one can come up with the expression for $T^{\mu\nu}$ in any frame: $$ T^{\mu\nu} = w u^\mu u^\nu - Pg^{\mu\nu},\quad(1.3) $$
where $ w = e + P$ is the heat function per unit volume.

As in the rest frame $\Gamma=1$, $u^0 = 1$, $u^i = 0$, and $g^{\mu\nu} = \mathrm{diag}(1, -1, -1, -1)$, we immediately recover $(1.2)$.



## 1.2 Definition of Basic Fluid Quantities

We need to introduce the notation for the basic quantities of the fluid before starting our derivation. <br>
The mass density $\rho$ and number density $n$ are related by
$$\rho = m n$$

By seperating the rest mass contribution, the previously introduced total energy density $e$ can be rewritten as
$$e = \rho c^2 + \rho \varepsilon,\quad(1.4)$$
with $\varepsilon$ being defined as the specific internal energy.
We also define the density of internal energy as $u = \rho \varepsilon$.

With this, we can rewrite heat function per unit volume $w$ (also called density of enthalpy) as
$$w = \rho h = \rho c^2 + u + P,\quad(1.5)$$
where we define the specific enthalpy as $h = c^2 + \varepsilon + \frac{P}{\rho}$.

Furthermore, we assume a perfect fluid with $$P = (\gamma - 1)u,\quad(1.6)$$ where $\gamma$ is the adiabatic index. This leads to the relations
$$P = (\gamma -1)\rho \varepsilon$$
$$h = c^2 + \gamma \varepsilon$$
$$w = \rho c^2 + \gamma \rho \varepsilon$$

From here on out, we will adopt the convention of setting $c=1$ throughout the derivations.

## 1.3. The equations of relativistic fluid dynamics


The equations of energy and momentum conservation are contained in
$$\partial_\mu T^{\mu}_\nu = 0 \quad(1.7)$$

One adds the equation of particle number conservation which reads as
$$\partial_\mu (nu^\mu) = 0 \quad(1.8)$$

From $(1.7)$ we obtain
$$ \partial_\mu T^{\mu}_\nu = u_\nu\partial_\mu(wu^\mu) + wu^\mu\partial_\mu u_\nu - \partial_\nu P = 0 \quad(1.9)$$

We now wish to write out these equations under the assumption of spherical symmetry.

1) Particle Number Conservation

Writing out Eq. (1.8) we get that
$$\partial_t(n\Gamma) + {\nabla} \cdot (n \Gamma \mathbf{v}) = 0 $$
Using spherical symmetry leads to ${\nabla} \cdot (n \Gamma \mathbf{v}) = \frac{1}{r^2} \partial_r (n\Gamma v)$, with $v$ being the radial component of $\mathbf{v}$,
and multiplying by the mass $m$, which is constant, we get that

$$ \frac{\partial (\rho \Gamma)}{\partial t} + \frac{1}{r^2}\frac{\partial (r^2 \rho \Gamma v)}{\partial r} = 0$$


2) Radial Component of Energy-Momentum Conservation

From Eq. (1.9) and paying attention to minus sign due to the lower index of $u_\nu$, we get

$$-\Gamma v \partial_t(w\Gamma) - \Gamma v \nabla \cdot (w\Gamma \mathbf{v}) - w(\Gamma \partial_t \Gamma v + \Gamma \mathbf{v} \cdot \nabla \Gamma v) - \partial_r P = 0$$

Using spherical symmetry gives $\mathbf{v} \cdot \nabla \Gamma v = v \partial_r \Gamma v$ and with the Leibniz rule, we can combine this result into

$$ \frac{\partial (w \Gamma^2 v)}{\partial t} + \frac{1}{r^2}\frac{\partial (r^2 w \Gamma^2 v^2)}{\partial r} = -\frac{\partial P}{\partial r} $$

3) Time Component of Energy-Momentum Conservation

Finally, starting from Eq. (1.9) once more we get that

$$\Gamma \partial_t (w\Gamma) + \Gamma \nabla \cdot (w\Gamma \mathbf{v}) + w(\Gamma \partial_t \Gamma + \Gamma \mathbf{v} \cdot \nabla \Gamma) -\partial_t P = 0$$

Using spherical symmetry and the Leibniz rule analogously to the case above, we see that

$$ \frac{\partial (w \Gamma^2  - P)}{\partial t} + \frac{1}{r^2}\frac{\partial (r^2 w \Gamma^2 v)}{\partial r} = 0$$

## 1.4. Rankine-Hugoniot Jump Conditions




We have now shown that the equations of motion derived from the stress energy tensor in spherical symmetry take the form of
$$ \frac{\partial (\rho \Gamma)}{\partial t} + \frac{1}{r^2}\frac{\partial (r^2 \rho \Gamma v)}{\partial r} = 0\quad(1.10a)$$

$$ \frac{\partial (\rho h \Gamma^2 v)}{\partial t} + \frac{1}{r^2}\frac{\partial (r^2 \rho h \Gamma^2 v^2)}{\partial r} = -\frac{\partial P}{\partial r} \quad(1.10b)$$

$$ \frac{\partial (\rho h \Gamma^2  - P)}{\partial t} + \frac{1}{r^2}\frac{\partial (r^2 \rho h \Gamma^2 v)}{\partial r} = 0 \quad(1.10c)$$

where the substitution $w=\rho h$ was made.

Given a discontinuity at $r = r_0$, these equations can be integrated to give the jump conditions in the shock frame. In particular, total spatial derivatives $\frac{\partial}{\partial r}$ with no prefactors will cause jump conditions.
Eq. (1.10a) can be written as

$$ \frac{\partial (\rho \Gamma)}{\partial t} + \frac{2}{r} \rho \Gamma v + \frac{\partial (\rho \Gamma v)}{\partial r} = 0$$

Integrating this about an arbitrarily small contour over the discontinuity will make the first two terms go to zero (both continuous) and the third one will give the jump condition (as it is a total derivative). This leads to the first jump condition  on $\rho \Gamma v.$ The other two follow in the same manner.

Regarding notation, quantities with a star $(*)$ are given in the shocked medium, quantities without a star are given in the unshocked medium. A prime $(')$ denotes the quantities in the comoving frame of the shock, while quantities without a prime are given in the lab frame.
$\Gamma$ and $\Gamma_*$ are the Lorentz factors of the unshocked/shocked media and $\Gamma_S$ is the Lorentz factor of the shock itself, all measured in the lab frame. This is also illustrated in fig. 1.1, where $\Gamma$ is already set to 1 as we will do it in the next section for strong shocks.

The three jump conditions are then given by

$$\rho_*' \Gamma_*' v_*' = \rho \Gamma' v' \quad(1.11a)$$

$$\rho h_* \Gamma_*'^2 v_*'^2 + P_* = \rho h \Gamma'^2 v'^2 + P\quad(1.11b)$$

$$\rho_* h_* \Gamma_*'^2 v_*' = \rho h \Gamma'^2 v'\quad(1.11c)$$

All quantities describing the fluid itself, such as pressure, energy density, ... are always given in the local comoving frame of the fluid.

Note here that dividing Eq. $(1.11c)$ by Eq. $(1.11a)$ gives $h_* \Gamma_*' = h \Gamma'$

## 1.5. Strong Shock Approximation

We will assume a strong shock, which leads to the conditions
$$v = 0, \Gamma = 1, P = 0  \text{ and } h = c^2 = 1$$ This implies that the lab frame and rest frame of the unshocked medium coincide.

$\gamma_*$ is the adiabatic index in the shocked medium and we have
$$P_* = (\gamma_* - 1)\rho_*\varepsilon_* \text{ and } h_* = 1 + \gamma_*\varepsilon_*$$

We will denote the rapidity of the different Lorentz factors as $\xi$. In the lab frame, we have $\xi_*$ for the shocked region, $\xi_S$ for the shock itself and $\xi = 0$ for the unshocked region. Since rapidities are additive, we get in the shock frame $\xi_*' = \xi_* - \xi_S$, $\xi_S' = 0$ and $\xi' = -\xi_S$.

Using $\Gamma = \cosh(\xi)$ and $\Gamma v = \sinh(\xi)$ as well as addition theorems for $\cosh$ and $\sinh$ gives

$$\Gamma_*' = \Gamma_* \Gamma_S - \Gamma_* \Gamma_S v_* v_S$$

$$\Gamma_*' v_*' = \Gamma_* \Gamma_S v_* - \Gamma_* \Gamma_S v_S$$

Also, from relative velocity addition formula we know $$v_*' = \frac{v_* - v_S}{1-v_* v_S}$$
Since the unshocked medium is at rest, there is no more difference between the comoving frame of the shock relative to the lab frame and the shockwave seen from the unshocked medium and we have $$v' = -v_S \text{ and } \Gamma' = \Gamma_S$$

Plugging all this information into the three jump conditions (1.11a-c) gives

$$\rho_* \Gamma_* \Gamma_S (v_* - v_S) = - \rho \Gamma_S v_S\quad(1.12a)$$

$$\rho_* (1 + \gamma_*\varepsilon_*) \Gamma_*^2 \Gamma_S^2 (v_* - v_S)^2 + (\gamma_* - 1)\rho_*\varepsilon_* = \rho h \Gamma_S^2 v_S^2 + P \quad(1.12b)$$

$$ \rho_* (1 + \gamma_*\varepsilon_*) \Gamma_*^2 \Gamma_S^2 (v_* - v_S)(1-v_*v_S) = -\rho h \Gamma_S^2 v_S \quad(1.12c)$$

From here, (1.12a) reads $$\rho_* = -\frac{\rho v_S}{\Gamma_* (v_S - v_*)}\quad(1.13)$$ This into (1.12c) gives us
$$1+\gamma_* \varepsilon_* = \frac{1}{\Gamma_* (1-v_* v_S)}\quad(1.14)$$

This can then be put into (1.12b) to eliminate $\varepsilon_*$ and a quick rearranging of terms leads to
$$\Gamma_S^2(\frac{v_* - v_S}{1-v_*v_S} + v_S) + \frac{(\gamma_* - 1)(1-\Gamma_* (1-v_*v_S))}{\Gamma_*^2(v_*-v_S)(1-v_*v_S)\gamma_*} = 0$$

Note now that $$\frac{v_* - v_S}{1-v_*v_S} + v_S = \frac{v_* (1-v_S^2)}{1-v_*v_S} = \frac{v_*}{\Gamma_S^2 (1-v_*v_S)}$$

This implies that
$$\Gamma_*^2(v_*-v_S)\gamma_* v_* = (1-\gamma_*)(1-\Gamma_* (1-v_*v_S))$$

Solving for $v_S$ and opening parantheses gives
$$v_S(\Gamma_* v_* (\gamma_* \Gamma_* - (\gamma_* - 1))) = \gamma_* \Gamma_*^2 v_*^2 - 1 + \gamma_* + \Gamma_* - \Gamma_* \gamma_*$$

Now, using that $\Gamma_*^2 v_*^2 = \Gamma_*^2 - 1$ leads to
$$v_S = \frac{(\Gamma_* - 1)(\Gamma_* \gamma_* + 1)}{\Gamma_* v_* (\gamma_* \Gamma_* - (\gamma_* - 1))}\quad(1.15)$$

From this the final simplified form of the jump conditions can be derived.

### 1) $\Gamma_S$ Jump Condition
Using $v_S$ from Eq.$\,$(1.15) gives
$$\Gamma_S^{-2} = 1 - v_S^2 = \frac{\Gamma_*^2 v_*^2 (\gamma_* \Gamma_* - (\gamma_* - 1))^2 - (\Gamma_* - 1)^2(\Gamma_* \gamma_* + 1)^2}{\Gamma_*^2 v_*^2 (\gamma_* \Gamma_* - (\gamma_* - 1))^2}$$

Just expanding the numerator puts it into the form
$$(\Gamma_* - 1)(\gamma_*(2-\gamma_*)(\Gamma_*-1)+2)$$ and by changing the brackets in the denominator we get

$$1-v_S^2 = \frac{\gamma_*(2-\gamma_*)(\Gamma_*-1)+2}{(\Gamma_* + 1)(\gamma_* (\Gamma_* - 1) + 1))^2}$$

### 2) $\rho_*$ Jump Condition
For this, consider first $\Gamma_* (1- \frac{v_*}{v_S})$ and substitute $v_S$ to get

$$\Gamma_* (1- \frac{v_*}{v_S}) = \frac{(\Gamma_* - 1)(\Gamma_* \gamma_* + 1)\Gamma_*  -\Gamma_*^2 v_*^2 (\gamma_* \Gamma_* - (\gamma_* - 1))}{(\Gamma_* - 1)(\Gamma_* \gamma_* + 1)}$$

Again, expanding the numerator gives $(\Gamma_* -1) (\gamma_* - 1)$ and using Eq.$\,$(1.13)
$$\frac{\rho_*}{\rho} = \frac{1}{\Gamma_*(1-\frac{v_*}{v_S})}$$ leads to

$$\frac{\rho_*}{\rho} = \frac{\Gamma_* \gamma_* + 1}{\gamma_* - 1}$$

### 3) $\varepsilon_*$ Jump Condition
Here, we first look at $\Gamma_*(1-v_*v_S)$ and again replace $v_S$ to get

$$\Gamma_*(1-v_*v_S) = \frac{\Gamma_* (\gamma_* \Gamma_* - (\gamma_* - 1)) - (\Gamma_* - 1)(\Gamma_* \gamma_* + 1)}{\gamma_* \Gamma_* - (\gamma_* - 1)}$$

The numerator, after expanding, simplifies to $1$ and we get
$$ \Gamma_*(1-v_*v_S) = \frac{1}{\gamma_* (\Gamma_* - 1) + 1} $$

Now using the expression for $\varepsilon_*$ from Eq.$\,$(1.14) and substituting this result gives
$$\varepsilon_* = \frac{1}{\gamma_*} \left( \frac{1}{\Gamma_*(1-v_*v_S)} - 1 \right) = \Gamma_* - 1$$

## 1.6. Simplified Strong Shock Jump Condition

We get three greatly simplified jump conditions in the strong shock approximation, which read

$$\varepsilon_* = \Gamma_* - 1\quad(1.16a)$$

$$\frac{\rho_*}{\rho} = \frac{\Gamma_* \gamma_* + 1}{\gamma_* - 1}\quad(1.16b)$$

$$\Gamma_S^2 = \frac{(\Gamma_* + 1)(\gamma_* (\Gamma_* - 1) + 1))^2}{\gamma_*(2-\gamma_*)(\Gamma_*-1)+2}\quad(1.16c)$$

**Non-relativistic limit**

In non-relativistic limit with $\Gamma_* =1 $, $\gamma_* = \frac{5}{3}, P_* = \frac{2}{3}\varepsilon_*$ we recover the famous sound shock-wave condition $ \rho_* = 4\rho $.

**Ultra-relativistic limit**

Further we work in approximation of ultrarelativistic shocked medium, which implies $ \gamma_* = \frac{4}{3},~P_* = \frac{1}{3}\varepsilon_*, $ and $\Gamma_* \gg 1.$

Assuming this, one contracts the system of jump conditions as follows:
$ (1.16c) $ goes to
$$ \Gamma_S^2 = \dfrac{\Gamma_*\cdot(\frac{4}{3}\,\Gamma_*)^2}{\frac{4}{3}\cdot(2 - \frac{4}{3})\cdot\Gamma_*} = \dfrac{\frac{4}{3}\cdot\frac{4}{3}}{\frac{4}{3}\cdot\frac{2}{3}}\Gamma_*^2. $$
Equation $ (1.16b) $ results into
$$ \dfrac{\rho_*}{\rho} = \dfrac{\frac{4}{3}\Gamma_*}{\frac{4}{3} - 1} = 4\Gamma_* $$

Thus, we arrive at a *surprisingly simple* set of equations:
$$ \varepsilon_* = \Gamma_*~~~~~~\quad(1.17a) $$
$$ \rho_* = 4\Gamma_*\rho~~\quad(1.17b) $$
$$ \Gamma_S = \Gamma_*\sqrt{2} \quad(1.17c) $$

# **2. Adiabatic expansion of a shock wave**

## 2.1. Energy conservation law

In order to treat an adiabatic expansion of a shock wave, we write the following energy conservation law [*in the lab frame*]
$$ \left(E_\mathrm{shock} + E_\mathrm{ext}\right)_{R_0} = \left(E_\mathrm{shock} + E_\mathrm{ext}\right)_{R} \quad(2.1)$$

In assumption of a strong shock and cold external matter one can use the following notation:

$$ E_\mathrm{shock}|_{R_0} = \Gamma_0 M_0 c^2, ~ E_\mathrm{ext}|_{R_0} = M_\mathrm{ext} c^2 \quad (2.2a) $$
$$ E_\mathrm{shock}|_{R} = \Gamma_* M_0 c^2, ~ E_\mathrm{ext}|_{R} = \Gamma' M_\mathrm{ext} c^2 \quad (2.2b) $$

We assume (initially) that all the medium behind the shock moves with the velocity corresponding to $\Gamma_*$. Hence, by Lorentz transformation along the r-axis we obtain $ \Gamma' = (\Gamma_*)^2 $ from
$$ \begin{pmatrix}
        \Gamma' M_\mathrm{ext} c^2 \\
        \Gamma' M_\mathrm{ext}v_r'
   \end{pmatrix} =
   \begin{pmatrix}
        \Gamma_* & -\beta_*\Gamma_* \\
        -\beta_*\Gamma_* & \Gamma_*
   \end{pmatrix}
   \begin{pmatrix}
        \Gamma_* M_\mathrm{ext} c^2 \\
        0
   \end{pmatrix} \quad(2.3) $$

Thus, the energy conservation law $(2.1)$ in notation $(2.2a,b)$ with $(2.3)$ can be seen as
$$ \Gamma_0 M_0 c^2 + M_\mathrm{ext} c^2 = \Gamma_* M_0 c^2 + (\Gamma_*)^2 M_\mathrm{ext} c^2 \quad(2.4a). $$

This allows to construct a dimensionless equation
$$ m\left(\dfrac{\Gamma_*}{\Gamma_0}\right)^2 + \left(\dfrac{\Gamma_*}{\Gamma_0}\right) - \left(1 + \dfrac{m}{\Gamma_0^2}\right) = 0, \quad(2.4b) $$
where $m = \Gamma_0\cdot M_\mathrm{ext} / M_0$.

## 2.2. Swept mass as an expansion parameter

Equation $(2.4b)$ is quadratic relative to $(\Gamma_*/\Gamma_0)$, hence

$$ \frac{\Gamma_*}{\Gamma_0} = \frac{-1 + \sqrt{1 + 4m\left(1 + \frac{m}{\Gamma_0^2}\right)}}{2m} \approx
\begin{cases}
1,~m\ll 1,\\
m^{-1/2},~1 \ll m \ll \Gamma_0^2,\\
\frac{1}{\Gamma_0} + \frac{\Gamma_0^2}{2m},~m\gg \Gamma_0^2
\end{cases}
\quad(2.5)
$$

the second root was omitted by straightforward physical reasoning ($\Gamma_* > 1$).

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp
from scipy.stats import trapezoid
```

```python
m = 10**np.linspace(-5, 10, 200)
gamma0 = 100

gamma_ratio = (-1 + np.sqrt(1 + 4*m*(1 + m/gamma0**2))) / (2*m)

plt.figure(figsize=(10, 4))
plt.title("Figure 2.1a. Deceleration of a shock wave as a function of swept up mass")

m_low = m[0:65]
m_mid = m[60:130]
m_high = m[110:]

plt.plot(m_low, gamma0 * np.ones_like(m_low), '--', color='red', label='ultrarelativistic')
plt.plot(m_mid, gamma0 * m_mid**(-1/2), '--', color='green', label='intermediate')
plt.plot(m_high, np.ones_like(m_high), '--', color='blue', label='newtonian 0th order')
plt.plot(m_high, 1 + gamma0**2/(2 * m_high), '--', color='blue', alpha=0.5, label='newtonian 1st order')

plt.plot(m, gamma0 * gamma_ratio, color='black', label='exact solution', linewidth=.7)

plt.xlabel(r'$m = \Gamma_0 (M_\mathrm{ext} / M_0)$, swept up mass')
plt.xscale('log')

plt.ylabel(r'$\Gamma$, shocked matter Lorentz-factor')
plt.yscale('log')

plt.grid(linestyle='--', color='lightgray')
plt.legend()

plt.tight_layout()
# plt.savefig("pictures/fig2_1a.png")
plt.show()
```

On fig. 1 one can see the decceleration of a shock wave as a function of the swept up mass.

The three stages of the decceleration process are clearly seen: almost constant-energy expansion at low swept mass range, rapid decelleration at intermediate swept-mass ranges, and *newtonian* stage corresponding to large swept-up masses.

Stage 1: $ M_\mathrm{ext} \ll \frac{M_0}{\Gamma_0} $

During this stage, the swept up mass is too low and no deceleration takes place.

Stage 2: $ M_\mathrm{ext} \approx \frac{M_0}{\Gamma_0} $

This threshold defines the beginning of deceleration and the Lorentz factor rapidly decays. In assumption of homogenous medium outside the considered object, one constructs a very simple relation between swept-up mass and radius:
$$ M_\mathrm{ext} = \dfrac{4\pi}{3}\pi\rho_0 R^3 \Longrightarrow R = \left(\dfrac{3M_\mathrm{ext}}{4\pi\rho_0}\right)^{1/3} $$
Using our condition that $ M_\mathrm{ext} \approx \frac{M_0}{\Gamma_0} $, we get a characteristic radius for when the deceleration starts
$$R_\mathrm{dec} = \left(\dfrac{3M_0}{4\pi\rho_0\Gamma_0}\right)^{1/3} = \left(\dfrac{3E_0}{4\pi n m_\mathrm{P}\Gamma_0^2 c^2}\right)^{1/3}$$
where we used that $E_0 = \Gamma_0 M_0 c^2$ and $\rho_0 \approx n m_\mathrm{P}$, neglecting electron mass and with $n$ being the nucleon number density.<br>
For typical parameters of GRB's, we get that $R_\mathrm{dec} \sim 10^{14}-10^{15}\,$m.


Stage 3: $ M_\mathrm{ext} \gg M_0 \Gamma_0 $

At this point, the ejected medium has decelerated enough to be considered as non-relativistic. The corresponding characteristic radius is by a factor of $\Gamma_0^{2/3}$ larger than $R_\mathrm{dec}$. We get
$$R_\mathrm{Newton} = \left(\dfrac{3E_0}{4\pi n m_\mathrm{P}}\right)^{1/3}$$

## 2.3. Swept-up mass - radius relation

Using the previously introduced characteristic radius $ R_\mathrm{0}^3 = \dfrac{3M_0}{4\pi\rho_0\Gamma_0}, $ we rewrite $(2.5)$ as
$$ \Gamma_*(r, \Gamma_0) =
\Gamma_0\cdot\frac{-1 + \sqrt{1 +4r^3\left(1 + \frac{r^3}{\Gamma_0^2}\right)}}{2r^3} \approx
\begin{cases}
        \Gamma_0(1 - r^3),~r\ll 1,\\
        \Gamma_0 r^{-3/2},~1 \ll r \ll \Gamma_0^{2/3},\\
        1 + \frac{\Gamma_0^2}{2r^3},~r\gg \Gamma_0^{2/3}
\end{cases}\quad(2.6) $$
using $ r = R/R_\mathrm{0} = m^{1/3}  $

```python
plt.figure(figsize=(10, 4))
plt.title("Figure 2.1b. Deceleration of a shock wave as a function of radius")

plt.plot(m_low**(1/3), gamma0 * np.ones_like(m_low), '--', color='red', label='ultrarelativistic')
plt.plot(m_mid**(1/3), gamma0 * m_mid**(-1/2), '--', color='green', label='deceleration')
plt.plot(m_high**(1/3), np.ones_like(m_high), '--', color='blue', label='newtonian 0th order')
plt.plot(m_high**(1/3), 1 + gamma0**2/(2 * m_high), '--', color='blue', alpha=0.5, label='newtonian 1st order')

plt.plot(m**(1/3), gamma0 * gamma_ratio, color='black', label='exact solution', linewidth=.7)

plt.xlabel(r'$r = R/R_0$, expansion radius')
plt.xscale('log')

plt.ylabel(r'$\Gamma$, shocked matter Lorentz-factor')
plt.yscale('log')

plt.grid(linestyle='--', color='lightgray')
plt.legend()

plt.tight_layout()
# plt.savefig("pictures/fig2_1ab.png")
plt.show()
```

## 2.4. Shock wave kinematics

The shock wave propagates with $\Gamma_S \approx \Gamma_*\sqrt{2}. $

**Remark.** *In the framework of one-zone model one cannot distinguish values $\Gamma_S$ and $\Gamma_*$. However, using $\Gamma_S$ leads to a clearly unphysical result: minimal velocity of the shock wave turns out to be $\approx 0.7c$. For that reason, in further calculations we use $\Gamma_*$ instead of $\Gamma_S$ to describe the shock wave evolution.*

As $ \Gamma_s = \left(1 - \dfrac{v_s^2}{c^2}\right)^{-1/2} = \left(1 - \dfrac{\dot{R}^2}{c^2}\right)^{-1/2}$, one can write the motion equation of a wave as a function of time.

From $(2.6)$ one finds

$$ \dfrac{dr}{dt} = \frac{c}{R_0}\sqrt{1 - \Gamma_*^{-2}(r, \Gamma_0)} \quad (2.7a) $$

It is convenient to rescale time and use a new dimensionless time variable $ \tau = \dfrac{t}{T_0}, $ where $ T_0 = R_0/c. $

***For typical values of a GRB, $T_0 \sim 1-2$ days.***

Hence, $ \dfrac{d}{d\tau} = \dfrac{\partial t}{\partial \tau}\dfrac{d}{dt} = T_0\dfrac{d}{dt} = \dfrac{R_0}{c}\dfrac{d}{dt}. $

And $(2.7a)$ becomes dimensionless:
$$ r_\tau := \dfrac{dr}{d\tau} = F(r; \Gamma_0) = \sqrt{1 - \Gamma_*^{-2}(r, \Gamma_0)} \quad (2.7b) $$

## 2.5. Analytical estimations for shock-wave dynamics

Approximate results in $(6)$ allow to solve the expansion equations analytically in the three limitng cases.

1. **Ultrarelativistic expansion**

    In the ballistic regime for low swept-up masses (and low radii) the shock wave's gamma-factor does not change significantly.

    From $(2.6)$ and $(2.7b)$
    $$ \Gamma_* \approx \Gamma_0; $$
    $$ r_\tau = \sqrt{1 - \Gamma_*^{-2}} \approx \sqrt{1 - \Gamma_0^{-2}} = \beta_0; $$
    $$ r_\tau = \beta_0 \tau \quad (2.8a) $$

2. **Intermediate stage**

    During the intermediate stage, when the swept-up mass is significant enough for the velocity to start dropping, but yet not large to stop the expansion $ 1 \ll m \ll \Gamma_0^2 $, from $(2.6)$ and $(2.7b)$ we get
    $$ \Gamma_* \approx \Gamma_0 r^{-3/2}; $$
    $$ r_\tau = \sqrt{1 - \Gamma_*^{-2}} \approx \sqrt{1 - r^3/\Gamma_0^2} \approx 1 - \frac{r^3}{2\Gamma_0^2} $$
    or in other words $\dfrac{dr}{1 - r^3/2\Gamma_0^2} = d\tau $ or $ \left(1 + r^3/2\Gamma_0^2\right)\,dr = d\tau $, which after integration gives
    $$ r + \dfrac{r^4}{8\Gamma_0^2} = \tau \quad (2.8b)$$
    In the first order approximation that can be inverted as
    $$ r \approx \tau - \dfrac{\tau^4}{8\Gamma_0^2} \quad(2.8b') $$

3. **Newtonian limit**

    In newtonian limit (when the swept mass is so large, that the shock wave hardly expands more, ($m\gg\Gamma_0^2$), from $(2.6)$ and $(2.7b)$ we get
    $ r_\tau = 0,~r = cst. $

    Numerical integration shows that this 0th order solution does not coincide with the desired results. So, we make the 1st order expansion by writing
    $ \Gamma_* = 1 + \dfrac{\Gamma_0^2}{2r^3} $

    $ r_\tau = \sqrt{1 - \Gamma_*^{-2}} \approx \sqrt{1 - 1 + \Gamma_0^2/r^3} = \Gamma_0^2/r^{3} $

    Integration leads to $ r = \left(\frac{5}{2}\Gamma_0\tau\right)^{2/5}\quad(2.8c) $



**Computational remark**

When using full expression $(2.6)$ to calculate the radius at low $r$, one gets significant computational errors, as for low $r$ we subtract $1$ from a number that is very close to $1$.

To smoothen the computational $\Gamma_*$, we decided to use an approximate expression $\Gamma_*\approx \Gamma_0(1 - r^3) + O(r^6)$ for $r < r_\mathrm{bnd} = 0.01. $

In Fig. 2.2. we show computational error (in red) and smoothened function (in blue), the boundary $\tau_\mathrm{bnd}$ is shown in green.

```python
def gamma_star_old(r0, gamma):
    return gamma * (-1 + np.sqrt(1 + 4*r0**3*(1 + r0**3/gamma**2))) / (2*r0**3)

def gamma_star(r0, gamma, r_bnd=0.01):
    res_correct = np.heaviside(r0 - r_bnd, 0.5) * gamma_star_old(r0, gamma)
    res_approx = np.heaviside(r_bnd - r0, 0.5) * gamma * (1 - r0**3)
    return res_approx + res_correct

def F(tau, r0, gamma):
    return np.sqrt(1 - (gamma_star(r0, gamma))**(-2))
```

## 2.6. Observer time as a measurable quantity

For a photon emitted from the distance $R$ from the center of the object, as the shock wave propagates with relativistic velocities, the observation time will differ from the time of the process.

Let the first photon $\gamma_1$ be emitted at time $(0, 0)$. After the time $ t $ (in spherical coordinates with the origin in the center of the compact object) it reached the point $ (t, ct)$.

At the same time, the second photon $\gamma_2$ is emitted from the shock wave front, which is now at the position $(t, R(t))$, which can be obtained from equation $(2.7)$.

Hence, despite the difference in the object-connected frame is $t$, a far observer will register these two photons with a time difference $t_\mathrm{obs}$, defined as
 $$ t_\mathrm{obs}(t) = t - \frac{R_0}{c} r(t) \quad (2.9a) $$

Or in dimensionliess values
$$ \tau_\mathrm{obs}(\tau) = \tau - r(t) \quad(2.9b) $$

## 2.7. Analytical expressions for the observer time

**1. Low masses (radii)**

$ (2.8a):\quad r \approx \beta_0\tau $

$ \tau_\mathrm{obs} \approx \tau - \beta_0\tau = \tau(1 - \beta_0) \approx \dfrac{\tau}{2\Gamma_0^2}$

$ r = \beta_0(1-\beta_0)\tau \quad(2.10a) $

**2. Intermediate masses (radii)**

$ (2.8b):\quad r + \dfrac{r^4}{8\Gamma_0^2} = \tau $

$ \tau_\mathrm{obs} \approx \dfrac{r^4}{8\Gamma_0^2} $

$ r = (8\Gamma_0^2\,\tau)^{1/4} \quad (2.10b) $

**3. Large masses (radii)**

$ (2.8c):\quad r = (2.5\Gamma_0\tau)^{2/5} $

$ \tau_\mathrm{obs} = \tau - (2.5\Gamma_0\tau)^{2/5} \approx \tau \quad (2.10c) $

# **3. Radiation analysis**

## 3.1 Relativistic Transformation Laws

Our aim is to describe an isotropically flashing sphere in its comoving frame and the flux recieved by a stationary observer far away. For this chapter, we denote the lab frame as $S$ and the comoving frame as $S'$, and consequently quantities without a prime are in the lab frame, quantites with a prime in the comoving frame. We first need to establish transformation laws of some important quantities. We keep $c=1$.

1) Frequency

We consider the observer being at rest, and an object moving at an angle $\theta$ with respect to the z-axis (measured in S), and corresponding Lorentz factor $\Gamma$. This object emits photons in the direction of the observer, so along the z-axis, see fig. 3.1 below. The time between two photon emissions in $S'$ is $t'$, and we know that $t = \Gamma t'$. <br>
During $t$, the object has moved by $\beta \cos\theta t$, causing the two photons to have a path difference of
$$\Delta s = \Gamma t' - \Gamma \beta \cos\theta t'$$
The observer recieves the two photons with a time difference of $\Delta t = \frac{\Delta s}{c} = \Delta s$. If we define
$$ D = \left[ \Gamma(1 - \beta\cos\theta) \right]^{-1}$$
and $\nu = \frac{1}{\Delta t}$, $\nu' = \frac{1}{t'}$, we arrive at the final result

$$\nu = \frac{\nu'}{\Gamma(1-\beta cos\theta)} = D\nu'\quad(3.1)$$

```python
plt.figure(figsize=(8, 6))
plt.title("Figure 3.1. Photon emitted from moving source")
plt.imshow(plt.imread("source-photon.png"))
plt.axis('off')
plt.show()
```

2)  Solid Angle

To determine the transformation of the solid angle $d\Omega$, we consider the following setup. We have the comoving frame $S'$ and the lab frame $S$ related by a boost along the z-axis with velocity $\beta$. <br>
In $S'$, a photon $\gamma'$ is emitted at an angle $\theta'$ to the z'-axis which in $S$ is seen as a photon $\gamma$ moving at an angle $\theta$ with respect to the z-axis, see also fig. 3.2. <br>
We denote the velocities of $\gamma$ and $\gamma'$ along the respective z-axis by $u_z$ and $u_z'$. From the relativistic velocity addition formula we know that
$$u_z = \frac{\beta + u_z'}{1 + \beta u_z'}$$
Using the initial setup, we also know that photons move with $c$, so $u_z = \cos\theta$ and $u_z' = \cos(\theta')$.
Plugging this in, we arrive at
$$ cos\theta = \frac{\cos(\theta') + \beta}{1 + \beta \cos(\theta')}\quad(3.2a)$$
Inverting this gives
$$cos(\theta') = \frac{\cos\theta - \beta}{1 - \beta \cos\theta}\quad(3.2b)$$

To get the transformation of the solid angle, note that $d\Omega = - d\varphi d\cos\theta$ and $d\Omega' = - d\varphi' d\cos(\theta')$. <br>
Since $d\varphi$ is orthogonal to the boost direction, it is unchanged and $d\varphi = d\varphi'$. <br>
For $d\cos(\theta')$ we use
$$d\cos(\theta') = \frac{d\cos(\theta')}{d\cos\theta} d\cos\theta$$

With (3.2b), we get that $$\frac{d\cos(\theta')}{d\cos\theta} = \frac{1}{\Gamma^2(1-\beta \cos\theta)^2} = D^2$$

and we finally get

$$ d\Omega' = D^2d\Omega\quad(3.3)$$

```python
plt.figure(figsize=(12, 4))
plt.title("Figure 3.2. Photon moving in frames S' and S")
plt.imshow(plt.imread("theta.png"))
plt.axis('off')
plt.show()
```

3) Total Energy

A source emits a number of photons instantaneously and isotropically in its comoving frame. We want to understand, using the previous analysis, how the total energy emitted by this burst is seen in the lab frame.

From assuming isotropic emission in the source's rest frame, we know that $\frac{dE'}{d\Omega'} = \frac{E_0'}{4\pi}$. Since $dE'$ for photons transforms as frequency, we can use the previous transformation laws for $dE'$ and $d\Omega'$ and get that
$$\frac{dE}{d\Omega} = D^3 \frac{dE'}{d\Omega'}\quad(3.4)$$

We can now compute the total emitted energy in the lab frame

$$E_0  = \int \frac{dE}{d\Omega} d\Omega = \int D^3 \frac{dE'}{d\Omega'} d\Omega = \frac{E_0'}{4\pi \Gamma^3} \int \frac{1}{(1 - \beta\cos\theta)^3} d\Omega = \frac{E_0'}{2 \Gamma^3} \int_0^\pi \frac{d\cos\theta}{(1 - \beta\cos\theta)^3} = \Gamma E_0' $$

So we arrive at the expected result of
$$ E_0 = \Gamma E_0'\quad(3.5) $$

## 3.2. Isotropic radiation

We now derived a very important property,
$$ \dfrac{dE}{d\Omega} = D^3(\Gamma, \theta)\dfrac{dE'}{d\Omega'},\quad D = \left[ \Gamma(1 - \beta\cos\theta) \right]^{-1}$$

The following figures show how this creates the characteristic cone shape in the lab frame.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def d_function(gm, th):
    beta = np.sqrt(1 - gm**(-2))
    return (gm * (1 - beta * np.cos(th)))**(-1)
```

```python
thetas = np.linspace(0, 2 * np.pi, 360)
gammas = [1, 2, 5]

plt.figure(figsize=(10, 6))
plt.subplot(1, 2, 1, projection='polar')
plt.title("Figure 3.3a. Boosted isotropic emission of photons\n\n")
colors = ['red', 'green', 'blue', 'orange', 'magenta', 'cyan']
for i, g in enumerate(gammas):
    f_dist_i = d_function(g, thetas)**3
    f_dist_i /= (max(f_dist_i))**0.9
    plt.plot(thetas, f_dist_i, label=r'$\Gamma = $'f'{g}', color=colors[i])
    plt.plot(1/g * np.ones(1000), 10**np.linspace(-10, np.log10(max(f_dist_i)), 1000), color=colors[i], alpha=.5)
    plt.plot(-1/g * np.ones(1000), 10**np.linspace(-10, np.log10(max(f_dist_i)), 1000), color=colors[i], alpha=.5)
    # plt.yscale('log')
plt.legend()

gammas = [1, 2, 4, 5]
plt.subplot(1, 2, 2, projection='polar')
plt.title("Figure 3.3b. Boosted isotropic for different gamma-factors\n\n")
for i, g in enumerate(gammas):
    f_dist_i = d_function(g, thetas)**3
    plt.plot(thetas, f_dist_i, label=r'$\Gamma = $'f'{g}', color=colors[i])
    plt.yscale('log')

plt.tight_layout()
plt.legend()

# plt.savefig("pictures/fig3_3.png")
plt.show()
```

# **4. Bolometric radiation**

## 4.1. Bolometric emission of a single flashing sphere

We consider our source to be a flashing sphere of radius $r_0$, so a number of sources homogenously distributed over the surface of a sphere emitting photons in the direction of the observer at a given time. We will express this quantity in terms of the observer time,
$$ t_\mathrm{obs} = t - \frac{r_0}{c}\cos\theta \quad(4.1a)$$
,with $t$ the time of the flash and $\theta$ the angle between z-axis and source located on the sphere. This can be cast into dimensionless form, as done in chapter 2.4 to get
$$ \tau_\mathrm{obs} = \tau - r_0\cos\theta \quad(4.1b)$$
From this, we get that $d\tau_\mathrm{obs} = -r_0d\cos\theta\quad(4.2)$.

We only want to consider the sources which are recieved at equal $\tau_\mathrm{obs}$. It is easy to see that this corresponds to a ring of the sphere at fixed $\theta$, determined by $\tau_\mathrm{obs}$. The number of sources on this ring is given by $dN = \sigma dA(\theta)$, $\sigma$ being the source surface density and
$dA = r_0^2 d\Omega = - 2 \pi r_0^2 d\cos\theta = 2\pi r_0 d\tau_\mathrm{obs}$.

So we get
$$dN = 2\pi r_0 \sigma d\tau_\mathrm{obs}\quad(4.3)$$

```python
plt.figure(figsize=(12, 6))
plt.title("Figure 4.1. Observed flux geometry of a flashing sphere")
plt.imshow(plt.imread("sphere-obs.png"))
plt.axis('off')
plt.show()
```

For the luminosity, we need to consider the ring of the sphere which has equal $\tau_\mathrm{obs}$, and thus by (4.1b) equal $\theta$. This ring has a luminosity of
$$\mathcal{L}= \frac{dE_{\mathrm{ring}}}{d\tau_\mathrm{obs}} =  \frac{E_\mathrm{source} dN}{d\tau_\mathrm{obs}} = 2\pi r_0 \sigma E_\mathrm{source} \quad(4.4)$$

Since this corresponds to isotropic radiation, a detector far away will only observe an amount of energy depending on its solid angle $\Omega_\mathrm{obs}$. This setup is illustrated in fig. 4.1. The infinitesimal amount of energy corresponding to $d\Omega_\mathrm{obs}$ is given by
$$dE = \frac{\partial \mathcal{L}}{\partial \Omega_\mathrm{obs}} d\Omega_\mathrm{obs} d\tau_\mathrm{obs} = 2 \pi r_0 \sigma \frac{\partial E_\mathrm{source}}{\partial \Omega_\mathrm{obs}}  d\Omega_\mathrm{obs} d\tau_\mathrm{obs}  $$

Just looking at the term $\frac{\partial E_\mathrm{source}}{\partial \Omega_\mathrm{obs}}$, we see that we can rewrite this as
$$\frac{\partial E_\mathrm{source}}{\partial \Omega_\mathrm{obs}} = D^3 \frac{\partial E_\mathrm{source}'}{\partial \Omega_\mathrm{obs}'} = D^3 \frac{E_0'}{4\pi} = D^3 \frac{E_0}{4\pi\Gamma}$$
where we used the transformation from Eq. (3.4).

Combining this gives
$$ dE = \frac{1}{2} r_0 \sigma D^3 \frac{E_0}{\Gamma} d\Omega_\mathrm{obs} d\tau_\mathrm{obs}\quad(4.5)$$

The observed flux is now defined as $\mathcal{F_\mathrm{obs}} = \frac{dE}{d\tau_\mathrm{obs} dS_{\perp}}$, where $dS_{\perp} = d^2 d\Omega_\mathrm{obs}$ and $d$ is the distance from the source to the observer. Plugging in we get

$$ \mathcal{F_\mathrm{obs}} = \frac{  \frac{1}{2} r_0 \sigma D^3 \frac{E_0}{\Gamma} d\Omega_\mathrm{obs} d\tau_\mathrm{obs}  }{  d^2  d\tau_\mathrm{obs}   d\Omega_\mathrm{obs}  } = \frac{1}{2} \frac{ 4 \pi r_0^2 \sigma E_0  D^3 }{ 4 \pi d^2 \Gamma r_0 } \quad(4.6)$$

Note that from the definitions we made, $4 \pi r_0^2 \sigma E_0 $ is the total number of sources multiplied by the total energy emitted per source. So this gives the total energy emitted by the flashing sphere, which we denote as $\mathcal{E}_0$.

We introduce two more definitions, which help us rewrite $1-\beta\cos\theta$, which appears in $D^3$. <br>
First, $\tau_\mathrm{obs,0} = \tau - r_0$ and second, $\Delta \tau_\mathrm{obs} = \frac{1-\beta}{\beta} r_0$. This can be combined into
$$ \frac{\tau_\mathrm{obs} - \tau_\mathrm{obs,0}}{\Delta \tau_\mathrm{obs}} = \frac{(1-\cos\theta)\beta}{1-\beta} $$
and we get that
$$(1-\beta\cos\theta) = (1-\beta)(1+ \frac{\tau_\mathrm{obs} - \tau_\mathrm{obs,0}}{\Delta \tau_\mathrm{obs}})\quad(4.7) $$

Substituting this into (4.6) gives

$$\mathcal{F_\mathrm{obs}} = \frac{1}{4 \pi d^2} \frac{ 1 }{ 2 \frac{\beta}{1-\beta}\Delta \tau_\mathrm{obs} (1-\beta)^3 \Gamma^4} \frac{ \mathcal{E}_0 }{ \left(1+ \frac{\tau_\mathrm{obs} - \tau_\mathrm{obs,0}}{\Delta \tau_\mathrm{obs}}\right)^3}
=  \frac{1}{4 \pi d^2} \frac{ (1+\beta)^2 }{ 2 \beta \Delta \tau_\mathrm{obs} (1-\beta^2)^2 \Gamma^4} \frac{ \mathcal{E}_0 }{ \left(1+ \frac{\tau_\mathrm{obs} - \tau_\mathrm{obs,0}}{\Delta \tau_\mathrm{obs}}\right)^3} $$

In the ultrarelativistic approximation, $\beta \approx 1$, and so we get the final result

$$\mathcal{F_\mathrm{obs}} \approx \frac{1}{4 \pi d^2} \frac{ 2 \mathcal{E}_0 }{\Delta \tau_\mathrm{obs} \left(1+ \frac{\tau_\mathrm{obs} - \tau_\mathrm{obs,0}}{\Delta \tau_\mathrm{obs}}\right)^3}$$

To summarize, we have a flashing sphere emitting photons of total energy $ \mathcal{E}_0 $ instantaneously.

We showed that the total emission of this sphere as a function of $ \tau_\mathrm{obs} $ can be written as
$$ \mathcal{F}_\mathrm{obs}(\tau_\mathrm{obs}; \mathcal{E_0}, \tau_\mathrm{obs,0}) = \dfrac{dE_\mathrm{obs}}{dS_\bot d\tau_\mathrm{obs}} = \dfrac{1}{4\pi d^2}\dfrac{2\mathcal{E}_0}{\Delta \tau_\mathrm{obs}}\left[ 1 + \frac{\tau_\mathrm{obs} - \tau_\mathrm{obs,0}}{\Delta \tau_\mathrm{obs}} \right]^{-3},\quad(4.8) $$
where $ \Delta \tau_\mathrm{obs} = \dfrac{1-\beta}{\beta}r_0 \approx \dfrac{r_0}{2\Gamma^2} $

```python
def flux(tau0, tau, r0, gamma, e0):
    beta = np.sqrt(1 - gamma**(-2))
    d_tau_obs = (1 - beta) / beta * r0
    return e0 / d_tau_obs * (1 + (tau - tau0) / d_tau_obs)**(-3)
```

## 4.2. Bolometric flux at a given observer time

For a sphere, flashing in space at time $t$ in the lab-frame, the infinitely far observer sees a *continuous* signal, which is given by (eqref).

The signal starts at $t_{{\mathrm{obs}}, 0}$, when the photons from the closest edge of the sphere reach the observer, and finishes at $t_{{\mathrm{obs}}, 0} + 2R/c$, when the photons from the opposite side of the sphere arrive.

The expansion of a shock-wave can be treated as a sequence of expanding flashing spheres. Thus, at a given time $t_{\mathrm{obs}}$, an observer on Earth detects the signals from all the spheres, flashed at $t\in [t_1, t_2]$ (in lab frame), such that the last photons from a sphere flashed at $t_1$ and the first photons from a sphere flashed at $t_2$ arrive at $t_\mathrm{obs}$.

In other words, $ t_{\mathrm{obs}} = t_{{\mathrm{obs}}, 0}(t_2) $ and $t_{\mathrm{obs}} = t_{{\mathrm{obs}}, 0}(t_1) + 2R(t_1)/c$.

The total bolometric flux then reads as
$$ \mathcal{F}_{\mathrm{obs}}(t_{\mathrm{obs}})
= \intop_{t_1}^{t_2}dt'\frac{\partial F_{\mathrm{obs}}^\mathrm{sph}(t_{\mathrm{obs}}; t')}{\partial t'}
= \dfrac{1}{4\pi d^2}\intop_{t_1}^{t_2}\dfrac{2L(t')dt'}{\Delta t_{\mathrm{obs}}(t')}\left(1 + \frac{t_{\mathrm{obs}} - t_{{\mathrm{obs}},0}(t')}{\Delta t_{\mathrm{obs}}(t')}\right)^{-3} \quad(4.9)$$

## 4.3. Radiated energy

We assume, that it is only the highly-relativistic matter right at the shock that radiates.
At time $t'$ (lab frame), the particles at the shock have energy $\Gamma_*M_0 c^2$. If only a (small) fraction $0 < \kappa \ll 1$ of this energy is irradiated, then
$$ L(t') = f\, M_0c^2\frac{d\Gamma_*}{dt}(t') $$

and the total bolometric flux changes to
$$ \mathcal{F}_{\mathrm{obs}}(t_{\mathrm{obs}})
= \dfrac{2\kappa M_0c^2}{4\pi d^2}\intop_{t_1}^{t_2}dt'\,\frac{d\Gamma_*}{dt}(t')\dfrac{1}{\Delta t_{\mathrm{obs}}(t')}\left(1 + \frac{t_{\mathrm{obs}} - t_{{\mathrm{obs}},0}(t')}{\Delta t_{\mathrm{obs}}(t')}\right)^{-3} $$

By changing integration variable from $t'$ to $\tau_0 = \tau_{\mathrm{obs}}(t')$, we rewrite
$$ f_{\mathrm{obs}}(\tau_{\mathrm{obs}}) = 2 \cdot\intop_{\tau_{{\mathrm{obs}},0}}^{\tau_{\mathrm{obs}}}d\tau_0\,\dfrac{d\Gamma_*}{d\tau_{\mathrm{obs}}}(\tau_0) \cdot \dfrac{1}{\Delta\tau_{\mathrm{obs}}(\tau_0)}\left(1 + \dfrac{\tau_{\mathrm{obs}} - \tau_0}{\Delta\tau_{\mathrm{obs}}(\tau_0)}\right)^{-3}, \quad (4.10)$$
where $\mathcal{F}_{\mathrm{obs}} = f_{\mathrm{obs}} \cdot \mathcal{F}_0 $ with $\mathcal{F}_0 = \kappa\dfrac{M_0c^3}{4\pi d^2 R_0}$, and $\tau_{{\mathrm{obs}},0} + 2r(\tau_{{\mathrm{obs}},0}) = \tau_{\mathrm{obs}}$.

##4.4. Flux approximation

Earlier by plotting the shape of one-flashing sphere spectrum we saw, that it has a large narrow peak at $\tau_{\mathrm{obs}}$ and then decays with the width $\Delta\tau$. One can think of substituting the $F_{\mathrm{obs}}^\mathrm{sph}$ with a properly-scaled Dirac delta-function.

This delta-like behavior is provided by this component
$$ \dfrac{1}{\Delta\tau_{\mathrm{obs}}(\tau_0)}\left(1 + \dfrac{\tau_{\mathrm{obs}} - \tau_0}{\Delta\tau_{\mathrm{obs}}(\tau_0)}\right)^{-3} \approx \tilde C\, \delta(\tau_{\mathrm{obs}} - \tau_0) $$

To find the proper scaling parameter, we integrate it by $\tau_{\mathrm{obs}}$ (as it is simpler) on $[\tau_{\mathrm{obs}}, +\infty)$.
$$ C = \intop_{\tau_{\mathrm{obs}}}^{+\infty}d\tau_{\mathrm{obs}}\,\dfrac{1}{\Delta\tau_{\mathrm{obs}}(\tau_0)}\left(1 + \dfrac{\tau_{\mathrm{obs}} - \tau_0}{\Delta\tau_{\mathrm{obs}}(\tau_0)}\right)^{-3} = \intop_{1}^{+\infty}\xi^{-3}d\xi = \dfrac{1}{2},$$
which allows to estimate the scaled bolometric flux as
$$ f_{\mathrm{obs}} \approx \frac{d\Gamma_*}{d\tau_{\mathrm{obs}}}(\tau_{\mathrm{obs}}) \quad (4.11)$$


## 4.5. Analytical estimations


From the previously obtained power laws in limiting cases, we can recalculate the resulting flux estimations.

**1. Ultrarelativistic regime**

$ r = \beta_0\tau \approx \tau,\quad \tau = 2\Gamma^2 \tau_{\mathrm{obs}} $

$ \Gamma_*\approx\Gamma_0(1 - r^3) = \Gamma_0\left(1 - \tau^3\right) = \Gamma_0\big(1 - 8\Gamma_0^{\,6}\,\tau_{\mathrm{obs}}^3\big) $

$ (f_{\mathrm{obs}})_\mathrm{UR}
= -d\Gamma_*/d\tau_{\mathrm{obs}}
= -24\cdot\Gamma_0^{\,7}\cdot\tau_{\mathrm{obs}}^2  \quad (4.12a)$

**2. Deceleration regime**

$ \tau_{\mathrm{obs}} \approx \dfrac{r^4}{8\Gamma_0^2} ~~ \Rightarrow ~~ r = \left(8\Gamma_0^2\tau_{\mathrm{obs}}\right)^{1/4}$

$ \Gamma_* \approx \Gamma_0 r^{-3/2} = \Gamma_0\cdot\left(8\Gamma_0^2\tau_{\mathrm{obs}}\right)^{-3/8} = 8^{-3/8}\Gamma_0^{1/4}\tau_{\mathrm{obs}}^{-3/8}$

$ (f_{\mathrm{obs}})_\mathrm{dec}
= -d\Gamma_*/d\tau_{\mathrm{obs}}
= -3\cdot8^{-11/8}\cdot\Gamma^{1/4}\cdot\tau_{\mathrm{obs}}^{-11/8}  \quad (4.12b) $

**3. Newtonian regime**

$ r = (2.5\Gamma_0\tau)^{2/5},\quad \tau_{\mathrm{obs}} \approx \tau ~~ \Rightarrow ~~ r = (2.5\Gamma_0\tau_{\mathrm{obs}})^{2/5}$

$ \Gamma = 1 + \Gamma_0^2/2r^3 = 1 + (2.5\Gamma_0\tau_{\mathrm{obs}})^{-6/5} \Gamma_0^2/ 2 = 0.5\cdot2.5^{-6/5}\cdot\Gamma_0^{-6/5}\tau_{\mathrm{obs}}^{-6/5}$

$ (f_{\mathrm{obs}})_\mathrm{Newt}
= -d\Gamma_*/d\tau_{\mathrm{obs}}
= -3\cdot2.5^{-6/5}\cdot\Gamma_0^{-6/5}\cdot\tau_{\mathrm{obs}}^{-11/5} \quad (4.12c) $

```python
from scipy.optimize import fsolve, brentq
from numpy import interp
from scipy.interpolate import interp1d

r = radius_values
t_obs = time_values - radius_values
gammas = gamma_star(radius_values, gamma0)
dg_dt =  np.diff(gammas) / np.diff(t_obs)

r_of_t_obs = interp1d(np.log(t_obs[1:]), np.log(r[1:]), kind='cubic')


def func(x, x0):
    return x + 2*np.exp(r_of_t_obs(np.log(x))) - x0

def integrand(t1, t_eval):
    y = flux(t_obs[1:], t_eval, radius_values[1:], gammas[1:], -dg_dt)
    return np.exp(interp(np.log(t1),np.log(t_obs[1:]), np.log(y)))

def flux_obs(t_eval):
    if (func(t_obs[1], t_eval) * func(t_obs[-1], t_eval)) >= 0:
        t_obs_min = 1e-15
    else:
        t_obs_min = brentq(func, t_obs[1], t_obs[-1], args=(t_eval,))
    t_obs_max = t_eval

    t_obs_line = 10**np.linspace(np.log10(t_obs_min), np.log10(t_obs_max), 10**4)
    res = trapezoid(integrand(t_obs_line, t_eval), t_obs_line)

    return res


fluxes_new = []
for i in range(0, len(t_obs)-1, 50):
    fluxes_new.append(flux_obs(t_obs[i+1]))

fluxes_new = np.array(fluxes_new)
```

# **5. Synchrotron radiation**

## 5.1. One-electron case

We assume that one electron emits as follows:

$$ P(\Gamma) = K_1B^2\Gamma^2~\mathrm{[W/electron]} \quad (5.1a) $$
$$ \nu_\mathrm{syn} = K_2B\Gamma^2~\mathrm{[Hz]} \qquad\qquad\quad~~~ (5.1b) $$

Single electron synchrotron power spectrum is given by

$$ P_\nu(\Gamma) = \dfrac{P(\Gamma)}{\nu_\mathrm{syn}}\Phi\left(\dfrac{\nu}{\nu_\mathrm{syn}}\right),\quad\Phi(\xi) = \dfrac{4}{3}
\begin{cases}
    \xi^{-1/3}, \xi \leqslant 1,\\
    0, \xi > 1.
\end{cases}\quad(5.2)$$

```python
def one_electron_radiation_curve(freq_i, scale = 1):
    freq = freq_i * scale
    return freq**(1/3) * np.heaviside(1 - freq, 0.5)

plt.figure(figsize=(6, 4))
plt.title("Figure 5.1. Synchrotron radiation curve")
f0 = 10**np.linspace(-3, 1.5, 1000)
plt.plot(f0, one_electron_radiation_curve(f0), color='royalblue', linewidth=3)
plt.xscale('log')
plt.xlabel(r'$\nu/\nu_\mathrm{syn}$, relative frequency')
plt.yscale('log')
plt.ylabel(r'$\Phi$, relative flux')
plt.grid(linestyle='dashed', color='lightgray')
plt.text(2e-2, 3.2e-1, "1/3", rotation=45)
# plt.savefig("pictures/fig5_1.png")
plt.show()
```

## 5.2. Naive electron distribution spectrum

Assume that the shocked medium electrons are the main source of radiation - the synchrotron radiation.

In the comoving frame of the shocked medium mass density of plasma is $\rho_*$, energy density is $\varepsilon_*$.

We define fractions $f_e$ and $f_B$ as follows:
$$ u_e = f_e \rho_* \varepsilon_*,\quad u_B = f_B \rho_* \varepsilon_* \quad (5.3)$$

Thus, for magnetic field one finds $u_B = B^2/8\pi \Longrightarrow B = \sqrt{8\pi u_B} = \sqrt{8\pi\,f_B\,\rho_*\varepsilon_*} = \sqrt{32\pi f_B \rho_\mathrm{ext}}\cdot\Gamma_* c$,

as $\varepsilon_* = \Gamma_*,~\rho_* = 4\Gamma_*\rho$

We also assume a power-law distribution of electrons in the shocked medium reference frame:
$$ n(\Gamma) = A\Gamma^{-s},  \quad (5.4)$$
where $2 < s < 3$ and $\Gamma > \Gamma_m$.

Now we are to investigate the spectrum $\mathcal{P}_\nu$ of the described electron distribution, where each particle radiates as $P_\nu$.

As $ d\mathcal{P}_\nu = P_\nu(\Gamma)dn = P_\nu(\Gamma)n(\Gamma)d\Gamma $, one writes

$\displaystyle\mathcal{P}_\nu = \intop_{\Gamma_m}^{+\infty}P_\nu(\Gamma)n(\Gamma)d\Gamma
= \intop_{\Gamma_m}^{+\infty}\dfrac{P(\Gamma)}{\nu_\mathrm{syn}}\Phi(\nu/\nu_\mathrm{syn})\cdot A\Gamma^{-s}d\Gamma = \intop_{\Gamma_m}^{+\infty}\dfrac{K_1 B^2 \Gamma^2}{K_2 B \Gamma^2}\Phi(\nu/K_2 B \Gamma^2)\cdot A\Gamma^{-s}d\Gamma = \\
\displaystyle = \left[\begin{matrix}
    \eta = \nu/K_2 B \Gamma^2 \to \Gamma = \sqrt{\nu/K_2B}\cdot\eta^{-1/2}\\
    d\Gamma = (-1/2)\sqrt{\nu/K_2B}\cdot\eta^{-3/2}d\eta
\end{matrix}\right] = \dfrac{1}{2}\dfrac{K_1BA}{K_2}\left(\dfrac{\nu}{K_2B}\right)^{-(s-1)/2} \intop_{0}^{\nu/\nu_m}\eta^{(s-3)/2}\Phi(\eta) d\eta$

**Case 1:** $\nu < \nu_m \Rightarrow \eta\in[0, 1) \Rightarrow \Phi(\eta) = \eta^{1/3}$, and the upper limit in the integral matters.

By integrating, $ \displaystyle\mathcal{P}_\nu = \dfrac{4}{3s - 1}\dfrac{K_1BA}{K_2}\Gamma_m^{-s+1}\left(\dfrac{\nu}{\nu_m}\right)^{1/3} \propto \nu^{1/3} $

**Case 2:** $\nu > \nu_m \Rightarrow$ upper limit can be set to 1, as above $\Phi = 0$.
$$ \mathcal{P}_\nu = \dfrac{4}{3s-1}\dfrac{K_1BA}{K_2}\Gamma_m^{-s+1}\left(\dfrac{\nu}{\nu_m}\right)^{-(s+1)/2} \propto \nu^{-(s+1)/2}$$

Finally,
$$ L_\nu = L_\mathrm{max}\begin{cases}
(\nu/\nu_m)^{1/3},~\nu < \nu_m,\\
(\nu/\nu_m)^{-(s-1)/2},~\nu \geqslant \nu_m.
\end{cases} \quad (5.5)$$

```python
def many_electron_radiation_curve(freq, s):
    p1 = freq**(1/3)*np.heaviside(1 - freq, 0.5)
    p2 = freq**(-(s-1)/2) * np.heaviside(freq - 1, 0.5)
    return p1 + p2

plt.figure(figsize=(6, 4))
plt.title("Figure 5.2. Radiation from a power-law electron distribution")
f0 = 10**np.linspace(-6.5, 3, 1000)
plt.plot(f0, many_electron_radiation_curve(f0, 2.5),
         color='orangered', linewidth=3)
plt.plot((1, 1), (6e-3, 1.5), color='royalblue', linestyle='dashed')

plt.text(5e-5, 5e-2, r'$1/3$', rotation=40)
plt.text(2e1, 4e-2, r'$-(s-1)/2$', rotation=-60)
plt.text(1.5, 6e-3, r'$\nu_m$')

plt.xscale('log')
plt.xlabel(r'$\nu/\nu_m$, relative frequency')
plt.yscale('log')
plt.ylabel(r'$\mathcal{P}$, spectral power')
plt.grid(linestyle='dashed', color='lightgray')

# plt.savefig("pictures/fig5_2.png")
plt.show()
```

## 5.3. Cooling energy

Curve $(5.5)$ gives an instant shape of the spectrum. However, electrons cool by radiating photons, and to obtain a correct spectrum, one has to take this into account.

To make an estimation, we introduce two timescales:

***Dynamical timescale*** $t_\mathrm{dyn}$ in the shockwave reference frame, which represents the time it takes for matter in the relativistic outflow to travel a distance comparable to the radius of the emission region
$$ t_\mathrm{dyn} = R(t) / \Gamma_* c$$

***Cooling timescale*** $t_\mathrm{cool}$ in the shockwave reference frame, which represents the time it takes for an electron to lose all its energy due to synchrotron radiation
$$ t_\mathrm{cool} = \dfrac{\Gamma m_e c^2}{P(\Gamma)} = \dfrac{m_e c^2}{K_1 B^2 \Gamma} $$

We define $\Gamma_c: t_\mathrm{dyn}(\Gamma_*) = t_\mathrm{cool}(\Gamma_c)$

$$ \dfrac{R}{\Gamma_*} = \dfrac{C_c}{\Gamma_*^2\Gamma_c} \quad ⇔ \quad \Gamma_c =  \dfrac{K_3}{r\,\Gamma_*} \quad(5.6)$$

In the beginning, $ \tau = 2\Gamma_*^2\tau_\mathrm{obs},~r \approx \tau = 2\Gamma_*^2 \tau_\mathrm{obs},~$ $\Gamma_C = K_3/2\Gamma_*^3\tau_\mathrm{obs},~$ which corresponds to the results from [Sari, Piran, Narayan].

So, if at $\tau_\mathrm{obs}$ $\Gamma_m > \Gamma_c$, all the electrons are quickly loosing energy due to synchrotron radiation, this situation will be further referred to as **fast cooling**.

If, on the contrary, at $\tau_\mathrm{obs}$ $\Gamma_m < \Gamma_c$, electrons with $\Gamma_e > \Gamma_c$ lose energy fast,  those electrons, whose $\Gamma_e < \Gamma_c$, can be considered as particles of constant energy, as synchrotron radiation losses are small. This regime is called **slow cooling**.

It turns out, that the synchrotron radiation spectra in this case are different. In the two following paragraphs we will show this difference.

Taking cooling into account changes the synchrotron radiation spectrum of a single electron: an electron with initial Lorentz factor $\Gamma > \Gamma_c$ cools down to $\Gamma_c$ in time $ t_c $. It means, that the cutoff frequency $\nu_\mathrm{syn}$ changes with time from $\nu_\Gamma = \nu_\mathrm{syn}(\Gamma)$ to $\nu_c = \nu_\mathrm{syn}(\Gamma_c)$.

We are interested in its average radiation power
$$ \langle P_\nu \rangle = \dfrac{1}{t_c}\intop_{0}^{t_c}dt\,P_\nu(\Gamma(t))$$

As stated earlier, $\nu_\mathrm{syn}(\tilde\Gamma) \propto \tilde\Gamma^2$, $ E_e(\tilde\Gamma) \propto \tilde\Gamma$, $P(\tilde\Gamma) \propto \tilde\Gamma^2$.

Hence, $ dE_e \approx -P(\Gamma)dt~\Rightarrow~dt \propto -d\Gamma,$ and

$$\langle P_\nu \rangle =
\dfrac{1}{t_c}\intop_{\Gamma_c}^{\Gamma}\dfrac{1}{P(\tilde\Gamma)}d\tilde\Gamma\left(\dfrac{\nu}{\nu_\mathrm{syn}(\tilde\Gamma)}\right)^{1/3}\mathbf{1}(\nu < \nu_\mathrm{syn}) \propto \nu^{1/3}\intop_{\Gamma_c}^{\Gamma_\nu}\dfrac{d\tilde\Gamma}{\tilde\Gamma^{2+2/3}}\propto \nu^{1/3}\Gamma_\nu^{-5/6}\propto\nu^{2/6 - 5/6} = \nu^{-1/2}$$

Which leads to $ P_\nu = P_\mathrm{max}\begin{cases}
(\nu/\nu_c)^{1/3},~\nu < \nu_c;\\
(\nu/\nu_c)^{-1/2},~\nu_c < \nu < \nu_\mathrm{syn};\\
0,~\nu > \nu_\mathrm{syn}.
\end{cases}\quad P_\mathrm{max} = P(\Gamma_c)\quad (5.7) $

```python
def single_electron_cooling_radiation_curve(freq, freq_c):
    p1 = (freq / freq_c)**(1/3) * np.heaviside(freq_c - freq, 0.5)
    p2 = (freq / freq_c)**(-1/2) * np.heaviside(1 - freq, 0.5) * np.heaviside(freq - freq_c, 0.5)
    return p1 + p2
```

```python
import matplotlib.patches as mpatches

plt.figure(figsize=(6, 4))
axs = plt.subplot(111)
plt.title("Figure 5.3. Synchrotron radiation from a cooling electron")
f0 = 10**np.linspace(-7, 2, 1000)

arrow = mpatches.FancyArrowPatch((5e-1, 0.02), (2e-4, 0.02),
                                 mutation_scale=80, alpha=.3,
                                 linewidth=0, color='purple')
axs.add_patch(arrow)
plt.text(1.3e-3, 0.018, "c o o l i n g", color='purple', fontsize=12)

plt.plot(f0, single_electron_cooling_radiation_curve(f0, 10**(-4)),
         color='orangered', linewidth=3, alpha=0.7)

plt.text(1e-6, 3e-1, r'$1/3$', rotation=40)
plt.text(1.5e-2, 6e-2, r'$-1/2$', rotation=-50)
plt.text(3e-5, 4e-3, r'$\nu_m$')
plt.text(1.2, 4e-3, r'$\nu_\mathrm{syn}(\Gamma)$')

for i in range(5):
  plt.plot(f0, one_electron_radiation_curve(f0, scale=10**i),
           alpha=0.4 + 0.15 * i, color='royalblue', linestyle='dashed')
plt.xscale('log')
plt.xlabel(r'$\nu/\nu_\mathrm{\Gamma}$, relative frequency')
plt.yscale('log')
plt.ylabel(r'$\Phi$, relative flux')
plt.grid(linestyle='dashed', color='lightgray')
# plt.savefig("pictures/fig5_3.png")
plt.show()
```

## 5.4. The fast cooling case

As in paragraph 5.2., $\displaystyle \mathcal{P}_\nu = \intop_{\Gamma_m}^{+\infty} P_\nu(\Gamma)n(\Gamma)d\Gamma $.

If $\nu < \nu_c:~$ $ \displaystyle \mathcal{P}_\nu = \intop_{\Gamma_m}^{+\infty} P_\mathrm{max}(\nu/\nu_c)^{1/3}A\Gamma^{-s}d\Gamma = N_e P_\mathrm{max}(\nu/\nu_c)^{1/3}$

If $\nu > \nu_c,~$ $ \displaystyle \mathcal{P}_\nu = \intop_{\Gamma_m}^{+\infty} P_\mathrm{max}(\nu/\nu_c)^{-1/2}\,\mathbf{1}(\nu < \nu_\mathrm{syn})\cdot A\Gamma^{-s}\,d\Gamma
= \left[\begin{matrix} \nu_\mathrm{syn} = K_2B\Gamma^2 = K\Gamma^2,~\Gamma = (\nu_\mathrm{syn}/K)^{1/2}\\
d\Gamma = \frac{1}{2}(\nu_\mathrm{syn}/K)^{-1/2}d\nu_\mathrm{syn} \end{matrix}\right] = \\ \displaystyle
= \frac{1}{2}P_\mathrm{max}(\nu/\nu_c)^{-1/2}\intop_{\nu_m}^{+\infty} \mathbf{1}(\nu < \nu_\mathrm{syn})\cdot A(\nu_\mathrm{syn}/K)^{-(s+1)/2}d\nu_\mathrm{syn}, $

where $\boldsymbol{1}(x\in A) = \begin{cases}1,\text{ if } x\in A,\\ 0,\text{ else}.\end{cases}$

Which is treated differently in two cases:

a) $ \nu_c < \nu < \nu_m $: $ \mathbf{1}(\nu < \nu_\mathrm{syn}) \equiv 1 $, and
$\mathcal{P}_\nu = N_e P_\mathrm{max}(\nu/\nu_c)^{-1/2}$;

b) $\nu > \nu_m$. In this case, the indicator function $\mathbf{1}(\nu < \nu_\mathrm{syn})$ cuts the lower limit at $\nu$, leading to
$ \displaystyle \mathcal{P}_\nu = \frac{1}{2}P_\mathrm{max}(\nu/\nu_c)^{-1/2}\intop_{\nu}^{+\infty} \mathbf{1}(\nu < \nu_\mathrm{syn})\cdot A(\nu_\mathrm{syn}/K)^{-(s+1)/2}d\nu_\mathrm{syn} = P_\mathrm{max}(\nu_m/\nu_c)^{-1/2}(\nu/\nu_m)^{-s/2}. $

The final version of the fast-cooling electrons spectrum is
$$\mathcal{P}_\nu = P_{\mathrm{max}}
\begin{cases}
    \left(\dfrac{\nu}{\nu_c}\right)^{1/3},~\nu < \nu_c, \\
    \left(\dfrac{\nu}{\nu_c}\right)^{-1/2},~\nu_c < \nu < \nu_m,\\
    \left(\dfrac{\nu_m}{\nu_c}\right)^{-1/2}\left(\dfrac{\nu}{\nu_m}\right)^{-s/2},~\nu > \nu_m.
\end{cases}\quad(5.8)
$$

```python
def fast_cooling_spectrum(g, nu, nu_c0, nu_m0, s=2.5):
    p1 = (nu / nu_c0)**(1/3) * np.heaviside(nu_c0 - nu, 0.5)
    p2 = (nu / nu_c0)**(-1/2) * np.heaviside(nu - nu_c0, 0.5) * np.heaviside(nu_m0 - nu, 0.5)
    p3 = (nu_m0 / nu_c0)**(-1/2) * (nu/nu_m0)**(-s/2) * np.heaviside(nu - nu_m0, 0.5)
    return (p1 + p2 + p3) * g
```

```python
s1 = 2.5
nu = 10**np.linspace(-3, 3, 1000)

plt.figure(figsize=(6, 4))
plt.title("Figure 5.4. Fast cooling electrons spectrum: "r"$\Gamma_m > \Gamma_c$")
nu_c = 0.1
nu_m = 10

plt.plot(nu, fast_cooling_spectrum(1, nu, nu_c, nu_m, s1),
         color='orangered', linewidth=3)

plt.text(5e-3, 5e-1, r'$1/3$', rotation=20)
plt.text(7e-1, 3e-1, r'$-1/2$', rotation=-25)
plt.text(1e2, 3e-3, r'$-s/2$', rotation=-50)

plt.plot((nu_c, nu_c), (1e-4, 1.5), color='royalblue', linestyle='dashed')
plt.text(0.5*nu_c, 1.2e-4, r'$\nu_c$')
plt.plot((nu_m, nu_m), (1e-4, 1.5), color='royalblue', linestyle='dashed')
plt.text(1.2*nu_m, 1.2e-4, r'$\nu_m$')

plt.xscale('log')
plt.yscale('log')
plt.grid(linestyle='dashed', color='lightgray')
# plt.savefig("pictures/fig5_4.png")
plt.show()
```

## 5.5. The slow cooling case


Here, as $\Gamma_m < \Gamma_c$, one can separate contributions from different components and split the integral over all $\Gamma$ values into two parts:
$ \displaystyle \int_{\Gamma_m}^{+\infty} = \underbrace{\int_{\Gamma_m}^{\Gamma_c}}_{cold} + \underbrace{\int_{\Gamma_c}^{+\infty}}_{hot} $.

In the first integral, all the electrons are 'cool', and have spectrum $(5.2)$. In the second integral, electrons are cooling, and have spectrum $(5.7)$

The 'cold' part produces the same spectral shape as $(5.5)$, the hot one has the same shape as the fast-cooling one, but with $\nu_m = \nu_c$.

By joining them and taking the leading contribution in each zone, we end up with the slow-cooling electrons spectrum

$$  \mathcal P_\nu = P_{\mathrm{max}}
\begin{cases}
    \left(\dfrac{\nu}{\nu_m}\right)^{1/3},~\nu < \nu_m, \\
    \left(\dfrac{\nu}{\nu_m}\right)^{-(p-1)/2},~\nu_m < \nu < \nu_c,\\
    \left(\dfrac{\nu_c}{\nu_m}\right)^{-(p-1)/2}\left(\dfrac{\nu}{\nu_c}\right)^{-p/2},~\nu > \nu_c
\end{cases}
$$

```python
def slow_cooling_spectrum(g, nu, nu_c0, nu_m0, s=2.5):
    p1 = (nu / nu_m0)**(1/3) * np.heaviside(nu_m0 - nu, 0.5)
    p2 = (nu / nu_m0)**(-(s-1)/2) * np.heaviside(nu - nu_m0, 0.5) * np.heaviside(nu_c0 - nu, 0.5)
    p3 = (nu_c0/nu_m0)**(-(s-1)/2) * (nu/nu_c0)**(-s/2) * np.heaviside(nu - nu_c0, 0.5)
    return (p1 + p2 + p3) * g
```

## 6.1. Electron properties evolution

In $(1.17a,b)$ we got $\varepsilon_* = \Gamma_*,~\rho_* = 4\Gamma_*\rho$, which implies $\epsilon_*\rho_* = 4\Gamma_*^2\rho$.

In $(5.3)$ we assumed that $ u_e = f_e \rho_* \varepsilon_*,~ u_B = f_B \rho_* \varepsilon_*, $ which now leads to $ u_e = 4f_e\Gamma_*^2\rho,~u_B = 4f_B\Gamma_*^2\rho $.

Electron spectrum shape $(5.4)$ gives $\displaystyle u_e = \intop_{\Gamma_m}^{+\infty} \Gamma m_e c^2\, A\Gamma^{-s} d\Gamma $

Thus, we reconstruct
$$ \Gamma_m = f_e\dfrac{s-2}{s-1}\dfrac{m_p}{m_e}\Gamma_* = C_m\Gamma_*\quad(6.1) $$

From $u_B = B^2/8\pi$ we obtain $ B = \sqrt{32\pi f_B\rho}\,\Gamma_* = C_B\Gamma_*$.

From $(5.6)$, substituting values for the constant, we obtain
$$ \Gamma_c = \dfrac{3m_e c^2}{16\sigma_T m_p}\dfrac{1}{n_\mathrm{ext}}\dfrac{1}{R\Gamma_*} = C_c\dfrac{1}{r\Gamma_*} \quad (6.2) $$

In $(5.1b)$ we stated that $ \nu_\mathrm{syn} = K_2B\Gamma^2$, which now turns to $\nu_\mathrm{syn} = C_2\Gamma_*\Gamma^2$.

```python
def find_min(value, x):
  min_ind = np.argmin(np.abs(value - x))
  min_val = x[min_ind]
  return min_ind, min_val
```

## 6.2. Spectrum of the source

All formulae of section 5 were derived in the comoving, and the corresponding primes were omitted there. Now we restore the primes, so by writing $P'_\nu$ we mean expressions $(5.8)$ or $(5.9)$, dependent on the corresponding regime.

In accordance with derivation of formula $(4.8)$, we write:

$\nu = D\nu',$ $E = D E',$ $dt = \Gamma dt'$, hence
$$ dE' = h\nu'\,dN'_\nu = h\nu' N p(\nu') d\nu'$$
$$ \dfrac{dE'}{d\Omega'd\nu'} = \dfrac{N}{4\pi} h\nu' p(\nu') $$
$$ \dfrac{dE}{d\Omega\,d\nu} = D^2\dfrac{dE}{d\Omega'd\nu'} = \dfrac{N}{4\pi}h\nu'p(\nu')D^2 $$
$$ dN^\mathrm{ring} = N\dfrac{dA^\mathrm{ring}}{4\pi R^2} = \dfrac{c dt_\mathrm{obs}}{2 R} $$
$$ L^\mathrm{ring}_{\mathrm{obs},\nu} = \dfrac{dE^\mathrm{ring}_\nu}{dt_\mathrm{obs}} = \dfrac{dN^\mathrm{ring}}{4\pi dt_\mathrm{obs}}h\nu'p(\nu')D^2 = \dfrac{cN}{4\pi R}h\nu'p(\nu')D^2$$
$$ F^\mathrm{ring}_{\mathrm{obs},\nu} = \dfrac{1}{4\pi d^2}\dfrac{cN}{2R}h\nu'p(\nu')D^2$$

Using again $\Delta t_\mathrm{obs},$ we get the spectral flux from a single sphere:
$$ F_{\mathrm{obs},\nu}(t_\mathrm{obs}) = \dfrac{2Nh\nu'p(\nu')}{4\pi d^2}\dfrac{1}{\Delta t_0}\left(1 + \dfrac{\tau_\mathrm{obs} - \tau_0}{\Delta \tau_0}\right)^{-2}$$

To obtain the flux from a sequence of flashing spheres we write as $(4.9)$
$$ \mathcal{F}_{\mathrm{obs},\nu} = \dfrac{2}{4\pi d^2}\intop_{t_1}^{t_2}dt \dfrac{dN}{dt} h\nu'p(\nu')\dfrac{1}{\Delta t_0}\left(1 + \dfrac{\tau_\mathrm{obs} - \tau_0}{\Delta \tau_0}\right)^{-2}$$

Now one can notice that $ dE'_\nu = h\nu' p(\nu') \dfrac{dN}{dt'} dt' $.

On the other hand, $ dE'_\nu = P'_\nu dt' $, hence $ P_\nu' = h\nu' p(\nu') \dfrac{dN}{dt'}, $ and

$$ \mathcal{F}_{\mathrm{obs},\nu} = \dfrac{2}{4\pi d^2}\intop_{t_1}^{t_2}dt \dfrac{dt'}{dt} P_\nu'(\nu/D)\dfrac{1}{\Delta t_0}\left(1 + \dfrac{\tau_\mathrm{obs} - \tau_0}{\Delta \tau_0}\right)^{-2} $$

and by changing variables and using that $ \frac{d\tau_\mathrm{lab}}{d\tau_\mathrm{obs}} = \dfrac{1}{1-\beta}\approx 2\Gamma $ we get
$$ \mathcal{F}_{\mathrm{obs},\nu} = \dfrac{4}{4\pi d^2}\intop_{\tau_0}^{\tau_\mathrm{obs}}d\tau P_\nu'(\nu/D)\dfrac{1}{\Delta\tau_0}\left(1 + \dfrac{\tau_\mathrm{obs} - \tau_0}{\Delta \tau_0}\right)^{-2},~D\approx2\Gamma\left(1 + \dfrac{\tau_\mathrm{obs} - \tau_0}{\Delta \tau_0}\right)^{-1}. \quad(6.3) $$

By using again delta-approximation and taking into account that $\displaystyle\intop_{\tau_0}^{+\infty}\dfrac{1}{\Delta\tau_0}\left(1 + \dfrac{\tau_\mathrm{obs} - \tau_0}{\Delta \tau_0}\right)^{-2}d\tau_\mathrm{obs} = 1$, we write
$$ \mathcal{F}_{\mathrm{obs},\nu} = \dfrac{4}{4\pi d^2}P_\nu'(\nu/2\Gamma) \quad(6.4) $$

From here it will be more convenient to work with observed frequencies $\nu_c = 2\Gamma \nu'_c$ and $\nu_m = 2\Gamma\nu'_m$.

*It is also more convenient to compare behavior with Sari & Piran article, as they provide these quantities in the observer-related frame.*

## 6.4. Bolometric flux comparison

Now one can compare bolometric fluxes, obtained in different methods:

**A simple bolometric flux calculation** $(4.10)$, that was based on the assumption of the constant radiative energy loss fraction of particles in the neighborhood of the shock wave. Scaling laws for the shock dynamics allowed us to calculate the power-law, which describes this curve in the deceleration region $(4.12b)$. We showed analytically that
$$ \mathcal{F}^\mathrm{bol}_\mathrm{obs} \propto \tau_\mathrm{obs}^{-11/8} $$

**Synchrotron radiation-based calculation** $(6.3)$ and $(6.4)$, that has been described in the present section.

For intermediate and high frequencies ($\nu \geqslant \nu_0$) in the deceleration region we see that the spectra in Figure 6.4. follow the power law
$$ F_\mathrm{obs}\propto\tau_\mathrm{obs}^{(2-3s)/4} $$

For the chosen value of power-law index $s = 2.5 = 5/2$ we surprisingly obtain
$$ (2 - 3\cdot(5/2))/4 = (4 - 15)/8 = -11/8\quad (!) $$

Which means, that in the integrated flux over a broad rangle of frequencies larger than $\nu_0$, which can be considered as bolometric flux, has the same scaling law in the deceleration region as our first naive calculation.

# **Summary**

In this project, we explored the physics of gamma-ray burst afterglows by modeling the kinematics of shock waves and deducing the emission properties of the burst during propagation.

We derived the fundamental equations of relativistic hydrodynamics and the Rankine-Hugoniot jump conditions, particularly for the case of a strong shock. This was then applied to the kinematics in a one-zone approach, showing the three characteristic phases of the evolution, ultra-relativistic regime, deceleration regime and newtonian regime.

Using the flashing sphere method, we were able to show the relativistic beaming shape of the emission and compute the observed bolometric flux in assumption of constant radiation losses of matter right at the shock front. This calculation shows that the observed bolometric flux has a maximum several seconds after the first photons from the gamma-ray burst arrive, and later, during the deceleration phase, the flux decreases as a power-law with a degree -11/8 for several days.

By combining adiabatic shockwave expansion dynamics and microphysical processes such as synchrotron radiation, we described the radiative processes of the shock and obtained the full spectral curve of emission. Our results show that high-energy spectrum peaks at the very first seconds after gamma-ray birst photons reaches the observer; the low-energy spectral component has a broad peak significantly later, several days after the gamma-rays.

Comparison of the integrated over intermediate and high-energy photons spectra shows that in the deceleration regions the integrated flux behaves as a power law of a degree -11/8, which coincides with bolometric flux calculations we did in the naive dynamical approach.

A natural next step towards improving these results would now be to abandon the one-zone model and derive an exact solution for the kinematics of the shock, as it was done by Blandford and McKee [3].
