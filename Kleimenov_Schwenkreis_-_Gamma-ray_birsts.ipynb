## 1. Relativistic hydrodynamics 

### 1.1. The energy-momentum tensor of a relativistic fluid

To construct the equations of relativistic hydrodynamics, one requires to know the energy-momentum tensor of a relativistic fluid.

The standard energy-momentum tensor, known from classical electrodynamics course, can be reads as following: 
$$ T = \begin{pmatrix}
    \varepsilon & p_x & p_y & p_z \\
    p_x & P & \sigma_{xz} & -\sigma_{xy} \\ 
    p_y & -\sigma_{xz} & P & \sigma_{xz} \\
    p_z & \sigma_{xy} & -\sigma_{xz} & P
\end{pmatrix},\quad(1) 
$$
where $\varepsilon$ is energy density, $\mathbf{p}$ is momentum-density, P is pressure and $ \hat\sigma $ corresponds to viscous forces tensor.

In a reference frame where the considered fluid is at rest, its momentum density $\mathbf{p} = 0$, and also the Pascal's law is valid, which reads $ T^{ij}d\mathbf{f}_i = Pd\mathbf{f}_j, $ and gives $T^{ij} = \delta^{ij}. $

Hence, in this very frame, the tensor $(1)$ simplifies to $(2)$.
$$ T = \begin{pmatrix}
    \varepsilon & 0 & 0 & 0 \\
    0 & P & 0 & 0 \\ 
    0 & 0 & P & 0 \\
    0 & 0 & 0 & P
\end{pmatrix}\quad(2) 
$$

Introducing the 4-velocity $u^\mu = (\gamma, \gamma\mathbf{v}/c)$, one can come up with the expression for $T^{\mu\nu}$ in any frame: $$ T^{\mu\nu} = w u^\mu u^\nu - Pg^{\mu\nu}.\quad(3), $$
where $w = \varepsilon + P$ is the heat function per unit volume.

As in the rest frame $\gamma=1$, $u^0 = 1$, $u^i = 0$, and $g^{\mu\nu} = \mathrm{diag}(1, -1, -1, -1)$, we immediately recover $(2)$.


### 1.2. The equations relativistic fluid dynamics

The equations of energy and momentum conservation are contained in
$$\partial_\mu T^{\mu}_\nu = 0. \quad(4)$$ 

One adds the equation of particle number conservation which reads as
$$\partial_\mu (nu^\mu) = 0. \quad(5)$$

From $(4)$ we obtain
$$ \partial_\mu T^{\mu}_\nu = u_\nu\partial_\mu(wu^\mu) + wu^\mu\partial_\mu u^\nu - \partial_\nu P = 0. \quad(6)$$

One can get the longitudal component by multiplying $(6)$ by $u^\nu$, and taking into account that $ u_\nu u^\nu = \gamma^2(1^2 - \mathbf{v}^2/c^2) = 1$, and hence $ u_\nu\partial_\mu u^\nu = 0 $:
$$ \partial_\mu(wu^\mu) - u^\mu\partial_\mu P = 0.\quad(7) $$

To consider the transversal component (in terms of $u^\mu$), we use the standard combination (8), which gives identical zero when multiplied by $u^\nu$:
$$ \partial_\mu T^\mu_\nu - u_\nu u^\mu\partial_\rho T^\rho_\mu = 0.\quad(8) $$

By a simple calculation, from $(8)$ we get
$$ wu^\mu\partial_\mu u_\nu = \partial_\nu P - u_\nu u^\mu\partial_\mu P $$

## 2. Adiabatic expansion of a shock wave

### 2.1. Energy conservation law

In order to treat an adiabatic expansion of a shock wave, we write the following energy conservation law [*in the lab frame*]
$$ \left(E_\mathrm{shock} + E_\mathrm{ext}\right)_{R_0} = \left(E_\mathrm{shock} + E_\mathrm{ext}\right)_{R} \quad(1)$$

In assumption of a strong shock and cold external matter one can use the following notation:

$$ E_\mathrm{shock}|_{R_0} = \Gamma_0 M_0 c^2, ~ E_\mathrm{ext}|_{R_0} = M_\mathrm{ext} c^2 \quad (2a) $$
$$ E_\mathrm{shock}|_{R} = \Gamma^* M_0 c^2, ~ E_\mathrm{ext}|_{R} = \Gamma' M_\mathrm{ext} c^2 \quad (2b) $$

We assume (initially) that all the medium behind the shock moves with the velocity corresponding to $\Gamma^*$. Hence, by Lorentz transformation along the r-axis we obtain $ \Gamma' = (\Gamma^*)^2 $ from
$$ \begin{pmatrix}
        \Gamma' M_\mathrm{ext} c^2 \\
        \Gamma' M_\mathrm{ext}v_r'
   \end{pmatrix} =
   \begin{pmatrix}
        \Gamma^* & -\beta^*\Gamma^* \\
        -\beta^*\Gamma^* & \Gamma^*
   \end{pmatrix}
   \begin{pmatrix}
        \Gamma^* M_\mathrm{ext} c^2 \\
        0
   \end{pmatrix} \quad(3) $$

Thus, the energy conservation law $(1)$ in notation $(2a,b)$ with $(3)$ can be seen as
$$ \Gamma_0 M_0 c^2 + M_\mathrm{ext} c^2 = \Gamma^* M_0 c^2 + (\Gamma^*)^2 M_\mathrm{ext} c^2 \quad(4a). $$

This allows to construct a dimensionless equation
$$ m\left(\dfrac{\Gamma}{\Gamma_0}\right)^2 + \left(\dfrac{\Gamma}{\Gamma_0}\right) - \left(1 + \dfrac{m}{\Gamma_0^2}\right) = 0, \quad(4b) $$
where $m = \Gamma_0\cdot M_\mathrm{ext} / M_0$.

### 2.2. Swept mass as an expansion parameter

Equation $(4b)$ is quadratic relative to $(\Gamma/\Gamma_0)$, hence
$$ \dfrac{\Gamma}{\Gamma_0} = \dfrac{-1 + \sqrt{1 + 4m(1 + \frac{m}{\Gamma_0^2})}}{2m} \approx
    \begin{cases}
        1,~m\ll 1,\\
        m^{-1/2},~1 \ll m \ll \Gamma_0^2,\\
        \Gamma_0,~m\gg \Gamma_0^2
    \end{cases}\quad(5)$$
the second root was omitted by straightforward physical reasoning ($\Gamma > 1$).

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp
```

```python
m = 10**np.linspace(-5, 10, 200)
gamma0 = 100

gamma_ratio = (-1 + np.sqrt(1 + 4*m*(1 + m/gamma0**2))) / (2*m)

plt.figure(figsize=(10, 4))
plt.title("Figure 1. Decceleration of a shock wave as a function of swept up mass")

m_low = m[0:80]
m_mid = m[60:130]
m_high = m[110:]

plt.plot(m_low, gamma0 * np.ones_like(m_low), '--', color='red', alpha=.5, label='ballistic expansion')
plt.plot(m_mid, gamma0 * m_mid**(-1/2), '--', color='green', alpha=.5, label='deceleration')
plt.plot(m_high, 1 * np.ones_like(m_high), '--', color='blue', alpha=.5, label='newtonian expansion')

plt.plot(m, gamma0 * gamma_ratio, color='royalblue', label='exact solution')

plt.xlabel(r'$m = \Gamma_0 (M_\mathrm{ext} / M_0)$, swept up mass')
plt.xscale('log')

plt.ylabel(r'$\Gamma$, shocked matter Lorentz-factor')
plt.yscale('log')

plt.grid(linestyle='--', color='lightgray')
plt.legend()

plt.tight_layout()
plt.show()
```

On fig. 1 one can see the decceleration of a shock wave as a function of the swept up mass.

The three stages of the decceleration process are clearly seen: almost constant-energy expansion at low swept mass range, rapid decelleration at intermediate swept-mass ranges, and *newtonian* stage corresponding to large swept-up masses.

Stage 1: $ M_\mathrm{ext} \ll M_0 $

## 2.3. Shock wave expansion dynamics

### Swept-up mass - radius relation

In assumption of homogenous medium outside the considered object, one constructs a very simple relation between swept-up mass and radius:
$$ M_\mathrm{ext} = \dfrac{4\pi}{3}\pi\rho_0 R^3 \Longrightarrow R = \left(\dfrac{3M_\mathrm{ext}}{4\pi\rho_0}\right)^{1/3} $$

By introducing the characteristic radius $ R_0^3 = \dfrac{3M_0}{4\pi\rho_0\Gamma_0}, $ we rewrite $(5)$ as
$$ \Gamma^*(r, \Gamma_0) = \Gamma_0\cdot\dfrac{-1 + \sqrt{1 +4r^3\left(1 + \frac{r^3}{\Gamma_0^2}\right)}}{2r^3} \quad(6) $$
using $ r = R/R_0 = m^{1/3}  $

```python
plt.figure(figsize=(10, 4))
plt.title("Figure 2. Decceleration of a shock wave as a function of radius")

plt.plot(m**(1/3), gamma0 * gamma_ratio, color='royalblue', label='exact solution')

plt.xlabel(r'$r = R/R_0$, expansion radius')
plt.xscale('log')

plt.ylabel(r'$\Gamma$, shocked matter Lorentz-factor')
plt.yscale('log')

plt.grid(linestyle='--', color='lightgray')
plt.legend()

plt.tight_layout()
plt.show()
```

#### Shock wave kinematics

The shock wave propagates with $\Gamma_S \approx \Gamma_*\sqrt{2}. $

As $ \Gamma_s = (1 - v_s^2/c^2)^{-1/2} = (1 - (\dot{R}^2)/c^2)^{-1/2}$, one can write the motion equation of a wave as a function of time.

From $(6)$ one finds

$$ \dot{r} = F(r; \Gamma_0) = \frac{c}{R_0}\sqrt{1 - \left[\Gamma_*(r, \Gamma_0)\sqrt{2}\right]^{-2}} \quad (7) $$

```python
def gamma_star(r, gamma):
    return gamma * (-1 + np.sqrt(1 + 4*r**3*(1 + r**3/gamma**2))) / (2*r**3)

def F(t, r, gamma):
    return np.sqrt(1 - (np.sqrt(2) * gamma_star(r, gamma))**(-2))
```

```python
# F1 = F(gamma0)
r1 = solve_ivp(F, (1e-3, 1e5), [1e-3], args=(gamma0,), t_eval=np.arange(1e-3, 1e5, 1e-1), method='Radau')

plt.figure(figsize=(10, 4))
plt.title("Figure 3. Radius-on-time dependence for a expanding shock wave")
plt.plot(r1.t, r1.y[0], color='royalblue', label='exact solution')
plt.grid(linestyle='--', color='lightgray')

plt.xlabel(r'$t$, time')
plt.xscale('log')
plt.yscale('log')
plt.ylabel(r'$r$, radius')
plt.show()
```

## 2.4. Observed time as a measurable quantity

For a photon emitted from the distance $R$ from the center of the object, as the shock wave propagates with relativistic velocities, the observation time will differ from the time of the process.

Let the first photon $\gamma_1$ be emitted at time $(t_0, 0)$. After the time $ \tau $ (in spherical coordinates with the origin in the center of the compact object) it reached the point $ (t_0 + \tau, c\tau)$.

At the same time, the second photon $\gamma_2$ is emitted from the shock wave front, which is now at the position $(t_0 + \tau, R(\tau))$, which can be obtained from equation $(7)$.

Hence, despite the difference in the object-connected frame is $\tau$, a far observer will register these two photons with a time difference $t_\mathrm{obs}$, defined as
 $$ c\tau_\mathrm{obs}(\tau) = c\tau - R_0 r(\tau) $$

```python
plt.figure(figsize=(10, 4))
plt.title("Figure 4. Radius on observed time dependence for a expanding shock wave")
plt.plot(r1.t - r1.y[0], r1.y[0], color='royalblue', label=r'$r(t_\mathrm{obs})$')
plt.plot(r1.t, r1.y[0], color='orange', label=r'$r(t)$', linestyle='dashed')
plt.grid(linestyle='--', color='lightgray')

plt.xlabel(r'$t$, time')
plt.xscale('log')
plt.yscale('log')
plt.ylabel(r'$r$, radius')
plt.legend()
plt.show()
```

# 3. Radiation analysis

## 3.1. Isotropic radiation

As the charged particles decelerate, they radiate. In the comoving reference frame of the shockwave, they radiate isotropically.

Now consider a point on a shockwave having radial velovity $v_s$ and moving at an angle $\theta_0$ to the direction to the observer (which we call $z$-axis).

This means that the isotropic radiation distribution $ \dfrac{dI}{d\Omega}\Bigg|_\mathrm{cmv} = \dfrac{4}{\pi} $ will be boosted along the r-axis, and the observer will see only the $ I(\theta_0) = \dfrac{dI}{d\Omega}\Bigg|_\mathrm{rest}(\theta=\theta_0)\,d\Omega_\mathrm{obs} $

The Lorentz-boost along the r-axis will be (in comoving cartesian coordinates)
$$ \Lambda =
\begin{pmatrix}
    \Gamma_S & 0 & 0 & -\beta_S \Gamma_S \\
    0 & 1 & 0 & 0 \\
    0 & 0 & 1 & 0 \\
    -\beta_S \Gamma_S & 0 & 0 & \Gamma_S \\
\end{pmatrix}. $$

Each point on a sphere is given by
$$ \vec{r} = \begin{pmatrix}
    x \\
    y \\
    z
\end{pmatrix} = \begin{pmatrix}
    \sin\theta\cos\varphi \\
    \sin\theta\cos\varphi \\
    \cos\theta
\end{pmatrix}
$$

Thus, we get
$$
\begin{pmatrix}
    \sin\theta\cos\varphi \\
    \sin\theta\cos\varphi \\
    \cos\theta
\end{pmatrix}_\mathrm{obs} =
\begin{pmatrix}
    0 & 1 & 0 & 0 \\
    0 & 0 & 1 & 0 \\
    -\beta_S \Gamma_S & 0 & 0 & \Gamma_S \\
\end{pmatrix}
\begin{pmatrix}
    0 \\
    \sin\theta\cos\varphi \\
    \sin\theta\cos\varphi \\
    \cos\theta
\end{pmatrix}_\mathrm{cmv} =
\begin{pmatrix}
    \sin\theta\cos\varphi \\
    \sin\theta\cos\varphi \\
    \Gamma_S\cos\theta
\end{pmatrix}_\mathrm{obs}
$$

# 4. Defining the radiating surface

For a point on a sphere (inclined on $\theta$) we get the following result:
$$ t_\mathrm{obs} = t - \frac{R(t)}{c}\cos\theta $$

Now we try to define all the points of the expanding system, from which one can get radiation at the fixed time $t_\mathrm{obs}$.

```python
from scipy.interpolate import interp1d
from scipy.optimize import root_scalar
```

```python
R0, C = 1, 1
```

```python
def get_the_times(i, t_values, r_values):
    thetas = np.linspace(0, 2 *np.pi, 180)  # an array of available angles

    t_obs = t_values[i] - R0 / C * r_values[i]
    print(f"t = {t_values[i]}, t_obs = {t_obs}")
    r_of_t = interp1d(t_values, r_values, kind='cubic')

    t_res = np.zeros([2, len(thetas)])
    for j, theta in enumerate(thetas):
        f_j = lambda t: t - t_obs - r_of_t(t) * np.cos(theta)
        if f_j(t_values[0])*f_j(t_values[-1]) > 0:
            t_res[:, j] = np.nan, theta
            continue
        sol_j = root_scalar(f_j, bracket=[t_values[0], t_values[-1]])
        t_res[:, j] = r_of_t(sol_j.root), theta
    return t_res
```

```python
plt.polar()
for i in [20, 100, 1000, 10000]:
    data = get_the_times(i, r1.t, r1.y[0])
    r, theta = data[0], data[1]

    plt.plot(data[1], np.log10(data[0]))
plt.show()
```
