# 1. Relativistic hydrodynamics

## 1.1. The energy-momentum tensor of a relativistic fluid

### Literature
[1] L. D. Landau, E. M. Lifschitz. *Fluid Mechanics - 2nd ed. - (Course of theoretical physics; v. 6)*. 1987. Pergamon Press.
[2] Л. Д. Ландау, Е. М. Лифшиц. *Гидродинамика - 4е изд. - (курс теоретической физики, т. 6)*. 2001. Физматлит.

To construct the equations of relativistic hydrodynamics, one requires to know the energy-momentum tensor of a relativistic fluid.

The standard energy-momentum tensor, known from classical electrodynamics course, can be reads as following: 
$$ T = \begin{pmatrix}
    \varepsilon & p_x & p_y & p_z \\
    p_x & P & \sigma_{xz} & -\sigma_{xy} \\ 
    p_y & -\sigma_{xz} & P & \sigma_{xz} \\
    p_z & \sigma_{xy} & -\sigma_{xz} & P
\end{pmatrix},\quad(1) 
$$
where $\varepsilon$ is energy density, $\mathbf{p}$ is momentum-density, P is pressure and $ \hat\sigma $ corresponds to viscous forces tensor.

In a reference frame where the considered fluid is at rest, its momentum density $\mathbf{p} = 0$, and also the Pascal's law is valid, which reads $ T^{ij}d\mathbf{f}_i = Pd\mathbf{f}_j, $ and gives $T^{ij} = \delta^{ij}. $

Hence, in this very frame, the tensor $(1)$ simplifies to $(2)$.
$$ T = \begin{pmatrix}
    \varepsilon & 0 & 0 & 0 \\
    0 & P & 0 & 0 \\ 
    0 & 0 & P & 0 \\
    0 & 0 & 0 & P
\end{pmatrix}\quad(2) 
$$

Introducing the 4-velocity $u^\mu = (\gamma, \gamma\mathbf{v}/c)$, one can come up with the expression for $T^{\mu\nu}$ in any frame: $$ T^{\mu\nu} = w u^\mu u^\nu - Pg^{\mu\nu}.\quad(3), $$
where $w = \varepsilon + P$ is the heat function per unit volume.

As in the rest frame $\gamma=1$, $u^0 = 1$, $u^i = 0$, and $g^{\mu\nu} = \mathrm{diag}(1, -1, -1, -1)$, we immediately recover $(2)$.


### 1.2. The equations relativistic fluid dynamics

The equations of energy and momentum conservation are contained in
$$\partial_\mu T^{\mu}_\nu = 0. \quad(4)$$ 

One adds the equation of particle number conservation which reads as
$$\partial_\mu (nu^\mu) = 0. \quad(5)$$

From $(4)$ we obtain
$$ \partial_\mu T^{\mu}_\nu = u_\nu\partial_\mu(wu^\mu) + wu^\mu\partial_\mu u^\nu - \partial_\nu P = 0. \quad(6)$$

One can get the longitudal component by multiplying $(6)$ by $u^\nu$, and taking into account that $ u_\nu u^\nu = \gamma^2(1^2 - \mathbf{v}^2/c^2) = 1$, and hence $ u_\nu\partial_\mu u^\nu = 0 $:
$$ \partial_\mu(wu^\mu) - u^\mu\partial_\mu P = 0.\quad(7) $$

To consider the transversal component (in terms of $u^\mu$), we use the standard combination (8), which gives identical zero when multiplied by $u^\nu$:
$$ \partial_\mu T^\mu_\nu - u_\nu u^\mu\partial_\rho T^\rho_\mu = 0.\quad(8) $$

By a simple calculation, from $(8)$ we get
$$ wu^\mu\partial_\mu u_\nu = \partial_\nu P - u_\nu u^\mu\partial_\mu P $$

# 2. Adiabatic expansion of a shock wave

## 2.1. Energy conservation law

In order to treat an adiabatic expansion of a shock wave, we write the following energy conservation law [*in the lab frame*]
$$ \left(E_\mathrm{shock} + E_\mathrm{ext}\right)_{R_0} = \left(E_\mathrm{shock} + E_\mathrm{ext}\right)_{R} \quad(1)$$

In assumption of a strong shock and cold external matter one can use the following notation:

$$ E_\mathrm{shock}|_{R_0} = \Gamma_0 M_0 c^2, ~ E_\mathrm{ext}|_{R_0} = M_\mathrm{ext} c^2 \quad (2a) $$
$$ E_\mathrm{shock}|_{R} = \Gamma^* M_0 c^2, ~ E_\mathrm{ext}|_{R} = \Gamma' M_\mathrm{ext} c^2 \quad (2b) $$

We assume (initially) that all the medium behind the shock moves with the velocity corresponding to $\Gamma^*$. Hence, by Lorentz transformation along the r-axis we obtain $ \Gamma' = (\Gamma^*)^2 $ from
$$ \begin{pmatrix}
        \Gamma' M_\mathrm{ext} c^2 \\
        \Gamma' M_\mathrm{ext}v_r'
   \end{pmatrix} =
   \begin{pmatrix}
        \Gamma^* & -\beta^*\Gamma^* \\
        -\beta^*\Gamma^* & \Gamma^*
   \end{pmatrix}
   \begin{pmatrix}
        \Gamma^* M_\mathrm{ext} c^2 \\
        0
   \end{pmatrix} \quad(3) $$

Thus, the energy conservation law $(1)$ in notation $(2a,b)$ with $(3)$ can be seen as
$$ \Gamma_0 M_0 c^2 + M_\mathrm{ext} c^2 = \Gamma^* M_0 c^2 + (\Gamma^*)^2 M_\mathrm{ext} c^2 \quad(4a). $$

This allows to construct a dimensionless equation
$$ m\left(\dfrac{\Gamma}{\Gamma_0}\right)^2 + \left(\dfrac{\Gamma}{\Gamma_0}\right) - \left(1 + \dfrac{m}{\Gamma_0^2}\right) = 0, \quad(4b) $$
where $m = \Gamma_0\cdot M_\mathrm{ext} / M_0$.

## 2.2. Swept mass as an expansion parameter

Equation $(4b)$ is quadratic relative to $(\Gamma/\Gamma_0)$, hence
$$ \dfrac{\Gamma}{\Gamma_0} = \dfrac{-1 + \sqrt{1 + 4m(1 + \frac{m}{\Gamma_0^2})}}{2m} \approx
    \begin{cases}
        1,~m\ll 1,\\
        m^{-1/2},~1 \ll m \ll \Gamma_0^2,\\
        \frac{1}{\Gamma_0} + \frac{\Gamma_0^2}{2m},~m\gg \Gamma_0^2
    \end{cases}\quad(5)$$
the second root was omitted by straightforward physical reasoning ($\Gamma > 1$).

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp
from scipy.stats import trapezoid
```

```python
m = 10**np.linspace(-5, 10, 200)
gamma0 = 100

gamma_ratio = (-1 + np.sqrt(1 + 4*m*(1 + m/gamma0**2))) / (2*m)

plt.figure(figsize=(10, 4))
plt.title("Figure 1. Decceleration of a shock wave as a function of swept up mass")

m_low = m[0:65]
m_mid = m[60:130]
m_high = m[110:]

plt.plot(m_low, gamma0 * np.ones_like(m_low), '--', color='red', label='ultrarelativistic')
plt.plot(m_mid, gamma0 * m_mid**(-1/2), '--', color='green', label='intermediate')
plt.plot(m_high, np.ones_like(m_high), '--', color='blue', label='newtonian 0th order')
plt.plot(m_high, 1 + gamma0**2/(2 * m_high), '--', color='blue', alpha=0.5, label='newtonian 1st order')

plt.plot(m, gamma0 * gamma_ratio, color='black', label='exact solution', linewidth=.7)

plt.xlabel(r'$m = \Gamma_0 (M_\mathrm{ext} / M_0)$, swept up mass')
plt.xscale('log')

plt.ylabel(r'$\Gamma$, shocked matter Lorentz-factor')
plt.yscale('log')

plt.grid(linestyle='--', color='lightgray')
plt.legend()

plt.tight_layout()
plt.show()
```

On fig. 1 one can see the decceleration of a shock wave as a function of the swept up mass.

The three stages of the decceleration process are clearly seen: almost constant-energy expansion at low swept mass range, rapid decelleration at intermediate swept-mass ranges, and *newtonian* stage corresponding to large swept-up masses.

Stage 1: $ M_\mathrm{ext} \ll M_0 $

## 2.3. Shock wave expansion dynamics

### Swept-up mass - radius relation

In assumption of homogenous medium outside the considered object, one constructs a very simple relation between swept-up mass and radius:
$$ M_\mathrm{ext} = \dfrac{4\pi}{3}\pi\rho_0 R^3 \Longrightarrow R = \left(\dfrac{3M_\mathrm{ext}}{4\pi\rho_0}\right)^{1/3} $$

By introducing the characteristic radius $ R_0^3 = \dfrac{3M_0}{4\pi\rho_0\Gamma_0}, $ we rewrite $(5)$ as
$$ \Gamma^*(r, \Gamma_0) = \Gamma_0\cdot\dfrac{-1 + \sqrt{1 +4r^3\left(1 + \frac{r^3}{\Gamma_0^2}\right)}}{2r^3} \approx
    \begin{cases}
        \Gamma_0(1 - r^3),~r\ll 1,\\
        \Gamma_0 r^{-3/2},~1 \ll r \ll \Gamma_0^{2/3},\\
        1 + \frac{\Gamma_0^2}{2r^3},~r\gg \Gamma_0^{2/3}
    \end{cases}\quad(6)$$
using $ r = R/R_0 = m^{1/3}  $

```python
plt.figure(figsize=(10, 4))
plt.title("Figure 2. Decceleration of a shock wave as a function of radius")

plt.plot(m_low**(1/3), gamma0 * np.ones_like(m_low), '--', color='red', label='ultrarelativistic')
plt.plot(m_mid**(1/3), gamma0 * m_mid**(-1/2), '--', color='green', label='deceleration')
plt.plot(m_high**(1/3), np.ones_like(m_high), '--', color='blue', label='newtonian 0th order')
plt.plot(m_high**(1/3), 1 + gamma0**2/(2 * m_high), '--', color='blue', alpha=0.5, label='newtonian 1st order')

plt.plot(m**(1/3), gamma0 * gamma_ratio, color='black', label='exact solution', linewidth=.7)

plt.xlabel(r'$r = R/R_0$, expansion radius')
plt.xscale('log')

plt.ylabel(r'$\Gamma$, shocked matter Lorentz-factor')
plt.yscale('log')

plt.grid(linestyle='--', color='lightgray')
plt.legend()

plt.tight_layout()
plt.show()
```

### Shock wave kinematics

The shock wave propagates with $\Gamma_S \approx \Gamma_*\sqrt{2}. $

**Remark.** *In the framework of one-zone model one cannot distinguish values $\Gamma_S$ and $\Gamma_*$. However, using $\Gamma_S$ leads to a clearly unphysical result: minimal velocity of the shock wave turns out to be $\approx 0.7c$. For that reason, in further calculations we use $\Gamma_*$ instead of $\Gamma_S$ to describe the shock wave evolution.*

As $ \Gamma_s = \left(1 - \dfrac{v_s^2}{c^2}\right)^{-1/2} = \left(1 - \dfrac{\dot{R}^2}{c^2}\right)^{-1/2}$, one can write the motion equation of a wave as a function of time.

From $(6)$ one finds

$$ \dfrac{dr}{dt} = \frac{c}{R_0}\sqrt{1 - \Gamma_*^{-2}(r, \Gamma_0)} \quad (7a) $$

It is convenient to rescale time and use a new dimensionless time variable $ \tau = \dfrac{t}{T_0}, $ where $ T_0 = R_0/c. $

Hence, $ \dfrac{d}{d\tau} = \dfrac{\partial t}{\partial \tau}\dfrac{d}{dt} = T_0\dfrac{d}{dt} = \dfrac{R_0}{c}\dfrac{d}{dt}. $

And $(7a)$ becomes dimensionless:
$$ r_\tau := \dfrac{dr}{d\tau} = F(r; \Gamma_0) = \sqrt{1 - \Gamma_*^{-2}(r, \Gamma_0)} \quad (7b) $$

### Analytical estimations

Approximate results in $(6)$ allow to solve the expansion equations analytically in the three limitng cases.

1. **Ultrarelativistic expansion**

    In the ballistic regime for low swept-up masses (and low radii) the shock wave's gamma-factor does not change significantly.

    From $(6)$ and $(7b)$
    $$ \Gamma_\star \approx \Gamma_0; $$
    $$ r_\tau = \sqrt{1 - \Gamma_\star^{-2}} \approx \sqrt{1 - \Gamma_0^{-2}} = \beta_0; $$
    $$ r_\tau = \beta_0 \tau \quad (8a) $$

2. **Intermediate stage**

    During the intermediate stage, when the swept-up mass is significant enough for the velocity to start dropping, but yet not large to stop the expansion $ 1 \ll m \ll \Gamma_0^2 $, from $(6)$ and $(7b)$ we get
    $$ \Gamma_\star \approx \Gamma_0 r^{-3/2}; $$
    $$ r_\tau = \sqrt{1 - \Gamma_\star^{-2}} \approx \sqrt{1 - r^3/\Gamma_0^2} \approx 1 - \frac{r^3}{2\Gamma_0^2} $$
    or in other words $\dfrac{dr}{1 - r^3/2\Gamma_0^2} = d\tau $ or $ \left(1 + r^3/2\Gamma_0^2\right)\,dr = d\tau $, which after integration gives
    $$ r + \dfrac{r^4}{8\Gamma_0^2} = \tau \quad (8b)$$
    In the first order approximation that can be inverted as
    $$ r \approx \tau - \dfrac{\tau^4}{8\Gamma_0^2} \quad(8b') $$

3. **Newtonian limit**

    In newtonian limit (when the swept mass is so large, that the shock wave hardly expands more, ($m\gg\Gamma_0^2$), from $(6)$ and $(7b)$ we get
    $ r_\tau = 0,~r = cst. $

    Numerical integration shows that this 0th order solution does not coincide with the desired results. So, we make the 1st order expansion by writing
    $ \Gamma_* = 1 + \dfrac{\Gamma_0^2}{2r^3} $

    $ r_\tau = \sqrt{1 - \Gamma_*^{-2}} \approx \sqrt{1 - 1 + \Gamma_0^2/r^3} = \Gamma_0^2/r^{3} $

    Integration leads to $ r = \left(\frac{5}{2}\Gamma_0\tau\right)^{2/5}\quad(8c) $



```python
def gamma_star_old(r0, gamma):
    return gamma * (-1 + np.sqrt(1 + 4*r0**3*(1 + r0**3/gamma**2))) / (2*r0**3)

def gamma_star(r0, gamma, r_bnd=0.01):
    res_correct = np.heaviside(r0 - r_bnd, 0.5) * gamma_star_old(r0, gamma)
    res_approx = np.heaviside(r_bnd - r0, 0.5) * gamma * (1 - r0**3)
    return res_approx + res_correct

def F(tau, r0, gamma):
    return np.sqrt(1 - (gamma_star(r0, gamma))**(-2))
```

```python
radius_values = 10**np.linspace(-3.5, -2.2, 3000)
plt.figure(figsize=(12, 4))
brd = 5e-3
gammas = gamma_star(radius_values, gamma0, r_bnd=brd)
gammas_old = gamma_star_old(radius_values, gamma0)

plt.title("Figure 3. Computational error of the Lorentz " + r"$\Gamma$-factor")
plt.plot(radius_values, gammas, 1)
plt.scatter(radius_values, gammas_old, 0.5, color='orange', alpha=.5)
plt.plot((brd, brd), (100-4e-5, 100+2e-5))
plt.ylim(100-4e-5, 100+2e-5)
plt.xlim(radius_values[0], radius_values[-1])
plt.xscale('log')
plt.show()
```

## 2.4. Observer time as a measurable quantity

### Definition of the observer time

For a photon emitted from the distance $R$ from the center of the object, as the shock wave propagates with relativistic velocities, the observation time will differ from the time of the process.

Let the first photon $\gamma_1$ be emitted at time $(0, 0)$. After the time $ t $ (in spherical coordinates with the origin in the center of the compact object) it reached the point $ (t, ct)$.

At the same time, the second photon $\gamma_2$ is emitted from the shock wave front, which is now at the position $(t, R(t))$, which can be obtained from equation $(7)$.

Hence, despite the difference in the object-connected frame is $t$, a far observer will register these two photons with a time difference $t_\mathrm{obs}$, defined as
 $$ t_\mathrm{obs}(t) = t - \frac{R_0}{c} r(t) \quad (9') $$

Or in dimensionliess values
$$ \tau_\mathrm{obs}(\tau) = \tau - r(t) \quad(9) $$

### Analytical expressions for the observer time

1. **Low masses (radii)**
    $$ (8a):\quad r \approx \beta_0\tau $$
    $$ \tau_\mathrm{obs} \approx \tau - \beta_0\tau = \tau(1 - \beta_0) \approx \dfrac{\tau}{2\Gamma_0^2}$$
    $$ r = \beta_0(1-\beta_0)\tau \quad(10a) $$
2. **Intermediate masses (radii)**
    $$ (8b):\quad r + \dfrac{r^4}{8\Gamma_0^2} = \tau $$
    $$ \tau_\mathrm{obs} \approx \dfrac{r^4}{8\Gamma_0^2} $$
    $$ r = (8\Gamma_0^2\,\tau)^{1/4} \quad (10b) $$
3. **Large masses (radii)**
    $$ (8c):\quad r = (2.5\Gamma_0\tau)^{2/5} $$
    $$ \tau_\mathrm{obs} = \tau - (2.5\Gamma_0\tau)^{2/5} \approx \tau \quad (10c) $$

```python
plt.figure(figsize=(14, 4))

# -----------
# FIRST PLOT
# -----------
plt.subplot(1, 2, 1)
plt.title("Figure 5a. Radius on observed time dependence")

t_obs = time_values - radius_values
plt.plot(t_obs, radius_values, color='black', label=r'$r(\tau_\mathrm{obs})$', linewidth=.7)
plt.grid(linestyle='--', color='lightgray')

# limiting cases
# low values
obs_low = t_obs[int(0.01 * n):int(0.25 * n)]
plt.plot(obs_low, beta0 * obs_low * 2 * gamma0**2, linestyle='--', color='red', label='ultrarelativistic')

# intermediate values
obs_mid = t_obs[int(0.15 * n):int(0.5 * n)]
plt.plot(obs_mid, (8 * gamma0**2 * obs_mid)**0.25, color='green', label='intermediate', linestyle='dashed')

# large values
obs_high = t_obs[int(0.35 * n):]
plt.plot(obs_high, (2.5 * gamma0 * obs_high)**0.4, color='blue', linestyle='dashed', label='newtonian')

plt.xlabel(r'$\tau_\mathrm{obs}$, observer time')
plt.xscale('log')
plt.yscale('log')
plt.ylabel(r'$r$, radius')
plt.legend()

# -----------
# SECOND PLOT
# -----------
plt.subplot(1, 2, 2)
plt.title("Figure 5b. Observed vs. real time")

plt.plot(time_values, t_obs, label='obs/real', color='black', linewidth=.7)

# limiting cases
# low values
time_low = time_values[:int(0.35 * n)]
plt.plot(time_low, time_low - beta0 * time_low, linestyle='--', color='red', label='ultrarelativistic')

# intermediate values
time_mid = time_values[int(0.2 * n):int(0.4 * n)]
plt.plot(time_mid, time_mid**4 / (8 * gamma0**2), color='green', label='intermediate', linestyle='dashed')

# large values
time_high = time_values[int(0.3 * n):]
plt.plot(time_high, time_high, color='blue', linestyle='dashed', label='newtonian')

plt.grid(linestyle='--', color='lightgray')
plt.xlabel(r'$t$, time')
plt.xscale('log')
plt.yscale('log')
plt.ylabel(r'$t_\mathrm{obs}$, radius')
plt.legend()

plt.show()
```

# 3. Radiation analysis

## 3.1. Isotropic radiation

**!!! Write down the derivation !!!**

Finally, we obtain the following important property:
$$ \dfrac{dE}{d\Omega} = \mathcal{D}^3(\Gamma, \theta)\dfrac{dE'}{d\Omega'},\quad\mathcal{D} = \left[ \Gamma(1 - \beta\cos\theta) \right]^{-1}$$



```python
def d_function(gm, th):
    beta = np.sqrt(1 - gm**(-2))
    return (gm * (1 - beta * np.cos(th)))**(-1)
```

```python
thetas = np.linspace(0, 2 * np.pi, 360)
gammas = [1, 2, 5]

plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1, projection='polar')
plt.title("Figure 7a. Boosted isotropic emission of photons")
colors = ['red', 'green', 'blue', 'orange', 'magenta', 'cyan']
for i, g in enumerate(gammas):
    f_dist_i = d_function(g, thetas)**3
    f_dist_i /= (max(f_dist_i))**0.9
    plt.plot(thetas, f_dist_i, label=r'$\Gamma = $'f'{g}', color=colors[i])
    plt.plot(1/g * np.ones(1000), 10**np.linspace(-10, np.log10(max(f_dist_i)), 1000), color=colors[i], alpha=.5)
    plt.plot(-1/g * np.ones(1000), 10**np.linspace(-10, np.log10(max(f_dist_i)), 1000), color=colors[i], alpha=.5)
    # plt.yscale('log')
plt.legend()

gammas = [1, 2, 4, 5]
plt.subplot(1, 2, 2, projection='polar')
plt.title("Figure 7b. Boosted isotropic for different gamma-factors")
for i, g in enumerate(gammas):
    f_dist_i = d_function(g, thetas)**3
    plt.plot(thetas, f_dist_i, label=r'$\Gamma = $'f'{g}', color=colors[i])
    plt.yscale('log')

plt.legend()
plt.show()
```

# 4. Bolometric radiation

## 4.1. The flashing sphere model

### Bolometric emission of a single flashing sphere

One can consider a flashing ball emitting photons of energy $ \mathcal{E}_0 $ instantaneously.

We showed that total emission of this ball as a function of $ \tau_\mathrm{obs} $ can be written as
$$ \mathcal{F}_\mathrm{obs}(\tau_\mathrm{obs}; \mathcal{E_0}, \tau_\mathrm{obs,0}) = \dfrac{dE_\mathrm{obs}}{dS_\bot d\tau_\mathrm{obs}} = \dfrac{1}{4\pi D^2}\dfrac{2\mathcal{E}_0}{\Delta \tau_\mathrm{obs}}\left[ 1 + \frac{\tau_\mathrm{obs} - \tau_\mathrm{obs,0}}{\Delta \tau_\mathrm{obs}} \right]^{-3}, $$
where $ \Delta \tau_\mathrm{obs} = \dfrac{1-\beta}{\beta}r_0 \approx \dfrac{r_0}{2\Gamma^2} $

```python
D = 1000  # in units of R0

def flux(tau, tau0, r0, gamma, e0):
    beta = np.sqrt(1 - gamma**(-2))
    dtau_obs = (1 - beta) / beta * r0
    return (4 * np.pi * D**2)**(-1) * 2*e0 / dtau_obs * (1 + (tau - tau0) / dtau_obs)**(-3)
```

```python
n0 = 0

plt.figure(figsize=(12, 4))
ax = plt.subplot(111)
for n in range(18, 100):
    tau_0 = t_obs[n0 + n * 100]
    tau_obs = t_obs[n0:]
    r_0 = radius_values[n0 + n * 100]
    gamma_0 = gamma_star(radius_values, gamma0)[n0]

    flux_1 = flux(tau_obs, tau_0, r_0, gamma_0, 1)
    ax.plot(tau_obs, flux_1)

ax2 = ax.twinx()

gammas = gamma_star(radius_values, gamma0)
ax2.plot(t_obs, gammas, color='black', label=r'Lorentz $\Gamma$-factor')

ax.set_title(r"Figure 8. Flux of flashing spheres at different $t_\mathrm{obs}$")
plt.grid(linestyle='--', color='lightgray')

ax.set_xlabel(r'$t_\mathrm{obs}$, time')
ax.set_xscale('log')

ax.set_ylabel(r'$F$, energy flux')
ax.set_yscale('log')
ax.set_ylim(1e-13, 1e17)

plt.legend()
plt.show()
```

### Convolution with the expansion of a shock wave

One can consider an observation of a shock wave as a succession of flashing spheres.

Then, the total observed flux will be
$$ \mathcal{F}_\mathrm{tot}(\tau_\mathrm{obs}) = \sum \mathcal{F}_i(\tau_\mathrm{obs}, \Delta\mathcal{E}_i, \tau_\mathrm{obs,0i})  $$

We assume that the energy value, carried by the sphere is proportional to kinetic energy loss of the medium right near the shock with $M_0$ and $ \Gamma_* $.

Hence, each sphere flashes the amount of energy $ \Delta\mathcal{E} = M_0 c^2 \Delta\Gamma_* = \dfrac{\Delta\Gamma_\star}{\Gamma_0} E_0 $.

And the sum becomes an integral:
$$ \mathcal{F}_\mathrm{tot}(\tau_\mathrm{obs}) = \dfrac{1}{4\pi D^2}\int\limits_{0}^{\tau_\mathrm{obs}} \dfrac{2d\Gamma/d\tau'}{\Delta \tau_\mathrm{obs}}\left[ 1 + \frac{\tau_\mathrm{obs} - \tau'}{\Delta \tau_\mathrm{obs}} \right]^{-3}\,d\tau' $$

```python
from scipy.integrate import simpson
```

```python
gammas = gamma_star(radius_values, gamma0)
dg_dt = np.diff(gammas[:]) / np.diff(time_values)

dt_obs = (radius_values / 2 * gammas)[1:]

fluxes = np.zeros(len(t_obs)-1)
t_prev = t_obs[0]
for i, t in enumerate(t_obs[1:]):
    if i == 0:
        continue
    fluxes[i] = simpson(flux(t, t_obs[1:], radius_values[1:], gammas[1:], -dg_dt[i]), t_obs[1:])
    t_prev = t
```

## 4.2. Alternative approach: defining the observed surface

For a point on a sphere (inclined on $\theta$) we get the following result:
$$ \tau_\mathrm{obs} = \tau - r(\tau)\cos\theta' $$

Now we try to define all the points of the expanding system, from which one can get radiation at the fixed time $t_\mathrm{obs}$.

```python
from scipy.interpolate import interp1d
```

```python
def get_the_times(i, t_values, r_values):
    thetas = np.linspace(0, 2 *np.pi, 180)  # an array of available angles

    t_obs = t_values[i] - r_values[i]
    print(f"t = {t_values[i]:.3f}, t_obs = {t_obs:.4f}")
    r_of_t = interp1d(t_values, r_values, kind='cubic')

    res = np.zeros([2, len(thetas)])
    for j, theta in enumerate(thetas):
        f_j = lambda t: t - t_obs - r_of_t(t) * np.cos(theta)
        if f_j(t_values[0])*f_j(t_values[-1]) > 0:
            res[:, j] = np.nan, theta
            continue
        sol_j = root_scalar(f_j, bracket=[t_values[0], t_values[-1]])
        res[:, j] = r_of_t(sol_j.root), theta
    return res
```

```python
plt.polar()
time_indices = [2000, 2500, 2800, 3000, 3500, 4000]
for i in time_indices:
    data = get_the_times(i, time_values, radius_values)
    r, theta = data[0], data[1]

    plt.plot(data[1], data[0])
    plt.yscale('log')
plt.show()
```
